# 1. Basics

Guide: https://github.com/SouravAggarwal/LLM-Deployment/blob/main/week2/day5.md

Automate Deployments with GitHub Actions
1. Actions: Run scripts in response to actions like doing a git push.
2. Workflow: A set of actions to run. YAML file stored in .github/workflows
3. Jobs (Execution): Collections of steps that run on a runner VM on Github.
