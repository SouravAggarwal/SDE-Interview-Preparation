```python
 # IEnumerable vs IQueryable
1.  IEnumerable
    - Execution in memeory and data is in memeory
    - Filtering: Happens after data is loaded
    - Namespace: System.Collections
    - Example:
        var data = db.Employees.ToList(); // Fetch all data
        var result = data.Where(e => e.Salary > 50000);

2. IQueryable
    - Remote execution and data is in external sources (DB)
    - Filtering: Happens at database level
    - namespace: System.Linq
    - Example:
        var result = db.Employees.Where(e => e.Salary > 50000); // Converts into a SQL query.
    - Example 2
        var result = employees
        .Where(e => e.Salary > 50000)
        .OrderBy(e => e.Name)
        .Select(e => new { e.Name, e.Salary });
    

# Linq Deferred execution
1. Execution is triggered by: ToList(), ToArray(), .First(), .Count() or any terminal operation
    - Deferred execution in LINQ means the query is not executed when defined, but only when the results are enumerated
    - Example:
        var numbers = new List<int> { 1, 2, 3, 4, 5 };
        var query = numbers.Where(n => n > 2); // Not executed yet
        foreach (var num in query) // Executed here
        {
            Console.WriteLine(num);
        }

# Select Many 
    - SelectMany → flatten collections

# async vs await
1. async
    - Marks a method async
    - Allows use of await inside the method
    - Changes the method return type to Task<T>
    

2. await
    - Pauses execution until the awaited task completes (without blocking the current thread)
    - When await is used, the method execution is paused until the task completes, but the underlying thread is not blocked and can be reused for other work. 
       - It does not necessarily create a new thread; instead, it relies on asynchronous I/O and task scheduling.

    - await does NOT create a thread. It works with OS-level async I/O operations and task schedulers, so OS notifies when response is ready
    - the await statement releases the current thread, and thread goes back to the thread pool.
    - After await is completed, UI operation might use the same thread, but backend apps might use any thread from threadpool.
    - It imporves scalability as it free the thread     
           
# Example
public async Task<string> GetData()
{
    await Task.Delay(2000); 
}

# Key Definations
Context → where the code runs (which thread/environment). Where should the code continue after await?”
State → what data the code uses (variables, parameters)
    
3. ConfigureAwait()
    - true (default, if not used): 
        - await SomeAsyncOperation().ConfigureAwait(true);
        - Captures the current context and Resumes on the same context (if one exists)
        - Example: mostly used in UI applications.
        - performance: slower
    - false
        - Do NOT capture the context and Resume on any available thread.
        - Example: mostly used in backend applications.
        - performance: faster
        - Key rule: always use false, except for UI application
               - What if we use false on UI application, The calling thread cannot access this object because a different thread owns it.
        
# Task vs Thread in C#
    1. Thread
        - A physical execution unit managed by the OS, Direct control over execution.
        - Direct OS scheduling
        - Dedicated resource  
        - Example:
            Thread t = new Thread(() =>
            {
                Console.WriteLine("Running in thread");
            });
            t.Start();

    2. Task
        - A logical unit of work (abstraction over threads)
        - Queued to ThreadPool
        - Example:
            Task.Run(() =>
            {
                Console.WriteLine("Running in task");
            });

# Task WhenAll
    - Task.WhenAll – Waits for all tasks to complete.
    - Task.WhenAny – Completes as soon as the first task finishes.

# Deadlock
    - Occurs when threads wait for each other indefinately.
    - Example: .Result() or .Wait()

# Kestrel
    - Kestrel is a cross-platform, high-performance web server for ASP.NET Core.

# Model Binding
    - Model Binding is the process of mapping HTTP request data (Json) to action method parameters automatically.

# NET Core vs .Net
 - .NET Framework → .NET Core → .NET 5+
 - .NET 8 version (cross platform, open source and C# version 12
 - When you run the code
    C# Code
       ↓
    Compiled to IL (Intermediate Language)
       ↓
    CLR executes using JIT compiler (CLR: Runtime that executes .NET code and manages memory) (JIT: Converts IL to machine code at runtime)
       ↓
    Machine code runs      

# ASP.Net Core
 - Framework for building web apps using .NET
 - Cross-platform, High performance, support Middleware, Routing, DI, Model Binding

Client (Browser/Postman)
        ↓
     Kestrel (Web Server)
        ↓
Middleware Pipeline
        ↓
Routing
        ↓
Controller
        ↓
Response                   

# Using
    - using ensures that an object is properly disposed after use.
     - It calls Dispose() in Finally block and run our code in try block.
     - Dispose releases unmanaged resources (like file handlers, http client)
     - Code
            using (var file = new StreamWriter("test.txt"))
            {
                file.WriteLine("Hello");
            }     
# Finalize
  - Called by Garbage collector                       
# Memory Leak
 - Memory leak occurs when memory is not released even though it is no longer needed.
 - Like using fileStream without using "using"

# Span
  - Does continous memeory allocation for list or similar struct, high performance.               

# class vs record
    - Class → reference type with identity (mutable by default)
    - Record → reference type with value-based equality (designed for immutability)  
        - Equality check: Record compares by value 
        - Example: var p1 = new Person("John"); var p2 = new Person("John");
                    Console.WriteLine(p1 == p2); // ✅ True      

        
```

```python
# Middleware
 - A chain of handlers through which every request passes.
  - It handles HTTP requests and responses in a pipeline, allowing processing such as authentication, logging, and error handling before passing control to the next component.
 - Client Request -> Middleware 1 -> Middleware 2 -> Controller -> return response.
 - Usecase
        - It can Inspect and modify each request/ response.
        - Built in Middleware: Authentication, Authorization, Logging, Exception handling.
 
            
            
# C# Example Code in Program.cs
            
var builder = WebApplication.CreateBuilder(args);

// 🔐 Secret key
var key = "THIS_IS_MY_SECRET_KEY_123456";

// Add Authentication
builder.Services.AddAuthentication(options =>
{
    options.DefaultAuthenticateScheme = JwtBearerDefaults.AuthenticationScheme;
    options.DefaultChallengeScheme = JwtBearerDefaults.AuthenticationScheme;
})
.AddJwtBearer(options =>
{
    options.TokenValidationParameters = new TokenValidationParameters
    {
        ValidateIssuer = false,
        ValidateAudience = false,
        ValidateLifetime = true,
        ValidateIssuerSigningKey = true,
        IssuerSigningKey = new SymmetricSecurityKey(Encoding.UTF8.GetBytes(key))
    };
});

builder.Services.AddAuthorization();
builder.Services.AddControllers();

// Middleware Pipeline
var app = builder.Build();
app.UseAuthentication();   // MUST come first, // 🔐 who are you? If request is not authorized, it can terminate and return response.
app.UseAuthorization();
app.MapControllers();
app.Run();

# Used in Controller
[ApiController]
[Route("api/[controller]")]
public class UserController : ControllerBase
{
    [HttpGet]
    [Authorize]
    public IActionResult Get()
    {
        var username = User.Identity.Name; // comes from JWT
        return Ok("You are authenticated!");
    }
}

 - here, No need to decode token manually, Middleware already did it.
```

```python
# Core C#
1. Value vs Reference
    - Value types (int, struct, enum) are stored on the stack and hold data directly.
    - Reference types (class, string, object) are stored on the heap; the variable holds a reference/pointer to the data.

2. Boxing vs Unboxing
    - Boxing: Converting a value type to object (stack → heap). object o = 42;
        - string can not boxed
    - Unboxing: Extracting the value type back from object. int i = (int)o;

3. == and .Equals()?
    - == Compares values or references
    - .Equals() compares value/ content.

4. readonly vs const?
    - const
        - Assign at Compile time
        - Static by default
    - readonly
        - Value assign at Runtime (only in constructor)
        - Value set per instance, only in constructor
    - static readonly
        - Value assign at Runtime (only in constructor)
        - Value shared across all instances, set once

5. String vs StringBuilder
 - string is immutable — every concatenation creates a new object.
 - StringBuilder is mutable — ideal for multiple string manipulations in loops

6. Overloading vs Overriding
    - Overloading: Same method name, different parameters (compile-time polymorphism).
    - Overriding: Redefining a base class virtual/abstract method in a derived class using override (runtime polymorphism).

7. virtual/ abstract/ override
    a. virtual
        - Has default implementation, can be overridden
    b. abstract
        - No implementation, must be overridden. Only in abstract class.
    c. override
        - Used in derived class to change behaviour

# Abstract example
 - abstract method only in abstract class
 - 
public abstract class Shape
{
    public abstract double GetArea(); // must override

    public virtual void Display()
    {
        Console.WriteLine("Shape info");
    }
}

public class Circle : Shape
{
    public double Radius { get; set; }

    public override double GetArea()
    {
        return Math.PI * Radius * Radius;
    }

    public override void Display()
    {
        Console.WriteLine("Circle");
    }
}


# Normal class and derived class
- Normal class uses virtual and override

public class Animal
{
    public virtual void Speak()
    {
        Console.WriteLine("Animal speaks");
    }
}

public class Dog : Animal
{
    public override void Speak()
    {
        Console.WriteLine("Dog barks");
    }
}

# Interface, with Interface no need to use any keyword
public interface ICalculator
{
    int Add(int a, int b);
    int Subtract(int a, int b);
}

public class Calculator : ICalculator
{
    public int Add(int a, int b)
    {
        return a + b;
    }

    public int Subtract(int a, int b)
    {
        return a - b;
    }
}
```
