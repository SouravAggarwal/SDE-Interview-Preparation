# 1. Basics

```python
AI: the science that empowers machines to learn from experiences, make decisions and perform human like tasks.
    
Generative AI (GenAI) = a branch of AI that can create new content (text, images, audio, video, code) that never existed befor and it looks like it was made by humans.

Core idea:
	1.	Learn a probability distribution of data (e.g., how words appear in sentences, how pixels form images).
	2.	Sample from that distribution to generate new things.

Technologies:
	•	Large Language Models (LLMs) → GPT, LLaMA (used for text/code)
	•	Diffusion Models → Stable Diffusion, DALL·E (used for images/video)
	•	GANs (Generative Adversarial Networks) → StyleGAN, DeepFakes (used for images/video)
	•	Transformers → Backbone for most GenAI (self-attention, sequence modeling).    
```

# 4. Terminologies

```python
1. Self Attention
     - Self-attention is a way for a model to figure out which words in a sentence are most important when trying to understand the meaning or make predictions.
     - Model assigns a weight to each word to decide how much attention it should pay to that word when understanding the sentence.   

2. Context Window
    -  The context window is the maximum number of tokens the Al model can process at a time, including both the input tokens (your prompt) and output tokens (the model's response).

3. Transformers
    • The transformer architecture is an Al model for processing text, designed to understand and generate language efficiently.
        1. Input Tokens (Words or Pieces of Words): It starts by breaking text into smaller chunks (tokens), like words or parts of words. For example, "I love cats" becomes [!, love, cats].
        2. Self-Attention (Focus Mechanism): Imagine reading a sentence like: The cat sat on the mat because it was soft.
            To figure out what "it" refers to, you need to think about "mat."
            The transformer uses self-attention to decide which words in the sentence are important to focus on. This helps it understand the context better.
        3. Layers (Like Thinking Steps):
            The model processes the input in multiple layers, where each layer refines its understanding. For example:
                • Layer 1 might focus on word meanings.
                • Layer 2 might figure out relationships between words.
                • Layer 3 might understand the whole sentence.
        4. Positional Information (Word Order):
            Since the model reads all the words at once, it also adds positional encoding to know the order of the words (e.g., knowing "the cat" is different from "cat the").
        5. Output (Final Prediction):
            After processing, it predicts the next word, translates text, summarizes, or performs another task. For example, if you type I love, , it might predict "cats."                       

4. Embeddings
    - It is the way of representing words, phrases, sentences, documents as dense vectors of numbers.
    - It capture the semantic meaning of text, allowing model to understand and work with language more effectively.
    - Input Tex: "Queen" => [4, 7, 8] (Vector Embedding)

5. Temperature
    - Temperature in natural language generation is a parameter that controls the randomness of predictions.                            
    - Temperature ranges from 0 and 2:
        High temperature introduces creativity and diversity. Sometime helusination as well.
        Low temperature ensures predictability and coherence.                                             

6. Token
    • A token is the smallest unit of text that an LLM processes.
    • 1 token is ~4 characters and or 1,000 tokens is ~750 words
	•	Depending on the tokenizer, a token can be:
    	•	A word (dog)
    	•	A subword (play + ing)
    	•	Just a few characters (hel + lo)
    	•	Or even punctuation (!, ,, )         

7.  Tokenizer?
	•	The tokenizer is the algorithm (and vocabulary mapping) that splits text into tokens and assigns each token an ID (integer).    
        Chat → 1234
```

# 6. Prompt Engineering

```python
Prompt engineering is the practice of designing prompts to get the best possible outputs from large language models (LLMs) like GPT-4.
Think of it like asking better questions to get better answers.
Good prompts can improve accuracy, creativity, reasoning, and efficiency.

a) Zero-Shot Prompting
	•	Just ask directly.
	•	Example:
	•	“Summarize this article in 3 bullet points.”

b) Few-Shot Prompting
    •	Few-shot prompting (in-context learning) involves providing a few example pairs of input and desired output to the generative Al model to guide its responses.
	•	Give examples of the desired output.
	•	Example:
        Translate English to French:
            - Dog → Chien
            - Cat → Chat
            - House → 

c) Chain-of-Thought Prompting
	•	Ask the model to “think step by step.”
    •	Chain-of-thought prompting is a technique that improves the reasoning abilities of Al models by breaking down complex questions or tasks into smaller manageable steps.
	•	Example:
    	•	“Solve this math problem step by step before giving the final answer.”
    - Example Few-Shot + Chain of Thought
        - Model Input
            • Q: Sarah has 10 pencils. She gives 4 pencils to her friend and buys a pack of 5 more pencils.
            How many pencils does Sarah have now?
            • A: Sarah started with 10 pencils. She gave away 4 pencils, leaving her with 6. Then she bought 5
            more pencils, so 6 + 5 = 11. The answer is 11.
            • Q: The library had 50 books. They lent 15 books to students and received 7 more books as a
            donation. How many books are in the library now?
        - Model Output
            • A: The library started with 50 books. They lent 15 books, so 50 - 15 = 35. Then, th
            more books, so 35 + 7 = 42. The answer is 42. received 7

d) Role Prompting
	•	Assign a role to the model.
	•	Example:
    	•	“You are a SQL expert. Write an optimized query for this schema…”

e) Instruction Refinement
	•	Be specific about format, length, and style.
	•	Example:
    	•	“Give me a JSON response with keys: name, age, skills.”   

f) Advanced Techniques
	•	Prompt Chaining → break a task into multiple prompts (pipeline).
	•	Self-Consistency → ask model to generate multiple solutions, pick the best.
	•	System + User Prompts → in APIs, define fixed system instructions + user query.
	•	Constraints & Guards → tell the model what NOT to do.

g) For Explaination
    •	Explain this algorithm like I’m 10 years old, then explain it like I’m a senior engineer.
        
```

```python
# Prompt Components

Your prompt should include:
1. Context: Background knowledge to help the model understand the
    - Its like assign a role to the model
    - Example: Assume that you are an expert financial analyst
2. Instructions: A clear statement fo the desired goal.
    - Example: Classify the following text as positive, neutral, or negative
3. Input Data: the data the model needs to process
    - Example: Text: "Apple stock increased by 2% in today's trading session"
4. Output Indicator: Tell the model, the format of output required.
    - Example: Sentiments
```

```python
System Prompt
    • These instructions define the overall behavior and tone of the Al system.
    • They control how the Al responds, shaping its role, tone, and communication style.
    • Example: "You are a helpful assistant who explains technical concepts in simple terms."

User Prompt:
    • User prompts are the questions or requests input by the user during the interaction.
    • They are dynamic and specific to the tas user seeks.
    • Example: "Explain the difference betwee and DC current."

Aspect        | Without System Prompt                              | With System Prompt
-----------------------------------------------------------------------------------------------
Behavior      | Generalized responses, context-independent.        | Role-specific responses based on the system prompt.
Tone          | Neutral, conversational, or simplified.            | Directed, consistent with the defined role or task.
Depth         | Broad and less detailed.                           | Detailed and precise, reflecting expertise or intent.
Focus         | Balances detail and simplicity for a wide audience.| Tailored to the specific context or domain.
Consistency   | Variable depending on user input alone.             | High consistency, adhering to the role or persona.
```

```python
# How to Improve Prompts, to get better Information
1. Multi-shot Prompting
2. Use of tools
3. Additional context (system message with background context, or better assign a role to the system.)
4. RAG
```

```python
Prompt Engineering Framework:
    
• Role: Specifies which role or capabilities the Al should assume.
• Task: Describes the concrete action or question posed to the Al.
• Context: Provides the Al with additional information to better perform the task.
• Goal: Defines what the outcome of the task should achieve.
• Constraint: Specifies any limitations or specific requirements for the task.
• Style: Sets the tone or mood of the desired output.
• Output Format: Determines how the result should be presented.

Good System Prompt:
 - Short, precise, the prompt should have clear understanding of how it should behave and when it should search the vector DB.
 - If system prompt is too long, the RAG application with slow down.
Example 1:
**Role:**
You are an assistant for my blog posts.

**Task:**
You write posts that exactly match my writing style. A sample is stored in the vector database.

**Info:**
You speak directly. Short, punchy sentences. Brutally honest - just like the sample in your uploaded knowledge.
```
