# 3. Quantization

```python
Quantization is the process of reducing numerical precision of a model’s parameters (weights, activations, sometimes gradients) to smaller data types to.
Its kind of a hack, that reduces the memory a lot, and reduces the performance accuracy by a little.

Istead of storing 32 bit number, we store 18/ 8/ 4 bit.
Pros:
    - Reduce model size, memory requirement, Inference latency.
    - Increases Throughput

# Size
Model Size (bytes) = Number of Parameters × Bytes per Parameter
FP32 (32 bits) = 32 / 8 = 4 bytes
Higher number means more accuracy, Its like light dimming switch, higher number means more combinations of switch.

- 7B parameters @ FP32 ≈ 28 GB
    - (7 * 10^9) * 4   => 28 GB
- 7B parameters @ INT8  ≈ 7 GB
- 7B parameters @ INT4  ≈ 3.5 GB    

# Binary Number
- 2^8 = 256 values (11111111)
- Either 0 to 155 or signed 8 bit (-128 to 128)

# Types of Training with Quantization
1. Post-Training Quantization
 - Train model -> Quantization -> Deploy
 - No retraining
 - Fast & Cheap
 - Slight accuracy drop

2. Quantization Aware Training 
 - Train model while simulating quantization
 - Higher accuracy
 - More training cost
 - More complex pipelines.
```
