# 2. Comparing Models

```python
# The Chinchilla Scaling Law
- Number of parameters ~ proportional to the number of training tokens.
- Assuming the model Architecture is same, If you increase your parameters by 2x, then you have to increase the training data by 2x as well.
- Althought, Recently we have found way to compress more information in the model with same parameters.

# Benchmark
Format: Benchmark (Whats being evaluated): Description

1. GPQA (PhD Science Expertise): 
    - 448 expert questions; non-PhD humans score ~34% even with web access
    - GPT-4 score 39% only, just above humans.

2. MMLU-PRO (Language understanding)
    - More advanced and cleaned-up version of MMLU(Massive Multitask Language Understanding), includes 10-choice questions instead of 4

3. AIME (Math)
    - Hard competitive math puzzles from a prestigious, invite-only competition for top high school students

4. LiveCode Bench (Coding)
    - Holistic benchmark for evaluating code LLMs using problems from LeetCode, AtCoder, and Codeforces. Questions are changed frequently.

5. MuSR (Reasoning) 
    - Logical deduction tasks, e.g., analyzing a 1,000-word murder mystery and answering who has means, motive, and opportunity like who has done the murder.
    - for reasoning models

6. HLE (Super-human intelligence)        
    - 2,500 extremely difficult, subject-diverse, multi-modal questions designed as a “final exam” for AI
    - This is the most complex test for humans.

# Limitation of Benchmarks
    - Training Data Contamination. Its difficult to keep the questions secretly.
    - Too narrow in scope
    - Saturation (Once model achieve good in resuts, then there is no point of having those questions)
    - Overfitting on the particular benchmark. 
```

```python
# Leaderboards
1. Artificial Analysis
    - The Leaderboard
    - https://artificialanalysis.ai/

2. Vellum
    - Includes context window, API costs and speed in one table.
    - https://www.vellum.ai/llm-leaderboard

3. Scale.com
    - SEAL Leaderborads: Good to choose a model for specific task.
    - https://scale.com/leaderboard#frontier

4. Hugging Face
    - Open Source Leaderboards, good for choosing model for a specific task (like medical, code)
    - https://huggingface.co/spaces?q=leaderboard                                                
    - https://huggingface.co/spaces/bigcode/bigcode-models-leaderboard                                                
    - Not updated regularly

5. Live Bench
    - Contamination free benchmark, as they update questions every 6 month.
    - https://livebench.ai/#/

# The Arena
- The LM Arena (earlier LMSYS) is an resource:
    - Compare Frontier and Open-source models directly via anonymous human votes
    - Blind human evals based on head-to-head comparison
    - LLMs are measured with an ELO-style score rating for results
    - Participate in the voting - it's a way to learn about different models while adding to the ecosystem
```
