# 1. Basics

```python
# Key Terms:
1. Latency: Time to process a single request
    - Analogy: How fast a car goes through a toll booth

2. Throughput: Number of requests per unit time
    - Ex: No. of requests per sec on server, No. of queries per second in DB.
    - Analogy: How many cars pass the booth per second
    - EX: Number of requests completed / second => 10 requests / 0.2s = 50 requests/sec

```

```python
# Calculation
1. Basics
    - 1 bit is used to store 0/1 only.
    - 4 bit can store (1011) => 2^4 = 16 distinct values or we can say it can store any number b/w range 0-15
    - 1 byte has 8 bits
    - 1 KB = 1024 Bytes
    - 1 MB = 1024 KB
    - 1 GB = 1024 MB
    - 1 TB = 1024 GB
    - 1 PB = 1024 PB

1. Text Characters
    - C sharp and Java uses UTF-16 encoding which takes 2 bytes/character.
    - 1 KB has 1024 bytes, so 1KB can store 1024/2 ~= 512 character of text.

2. Text with no. of users
    - 1 million user (10^6) * 1 KB => 10^3 * 10^3 * 1 KB => 1 GB
    - 1 billion user (10^9) * 1 KB => 10^3 * 10^3 * 10^3 * 1 KB => 1 TB or 0.931 TB to be precise
```

# 2. Framework for system design

```python
1. Decide Requirements
    1. Functional requirement: What should the system do?
        - Ex: youtube: users can upload and stream videos.
    2. Non-Functional requirements: How well does the system performs?
        - Reliability, Scalability, Maintainability
        - Ex: Uploading should not take ours, so latency matters, No.  of users it can handle, 

2. Capacity Estimation
    - Why this step is required?
        - How many servers, Databases, caches are required for our system?
        - Cost estimation?
        - Decide type and specification of all the hardwares?
            - Like type of DB?
        - It helps us to determine if our system is read heavy or write heavy?
            - For example, if read heavy then SQL with index, else if write heavy then casandra
                                                          
    1. DAU/ MAU (daily and monthly active users)
    2. Throughput
    3. Storage and memeory: Amount of memory required
    4. Network and Bandwidth: How much data moves in and out of system.
        - Ingress: Data flowing into our system
        - Outgress: Data flowing out from our system per second.

3. API Design
    1. Endpoint
        - the location where a service lives on the web
    2. Body
    3. Response

4. High Level Design
    - Draw diagram
        
```

# 3. Instagram news Feed

```python
1. Requirements
    1. Functional Requirement
        - Create Social Media Posts
        - Follow/ Unfollow 
        - News Feed of the users they follow (Newest first)
        - Like and comments
        - Notification

    2. Non-functional Requirement
        - Availability: High Availability (99.99% of uptime)
        - Eventual Consistancy: If a user posts something, its okay if it doesn't show up immediately but appears within a few seconds (1-2 seconds)
        - Low Latency: Home button should load all feeds in a second.
        - Scalability: 500 DAU
        - Extnesibility: Adding new features should be easy.
        - Usability: Text loads, but image is taking too much time (this affects user experience)

2. Capacity Estimation
    1. DAU: 500 million
    2. MAU: 2 billion
    3. Throughput
        1. Write Throughput
            - According to functional req: we write in 3 ways
                A. Creating posts (text, image, vide), Following, Like and comment
                    - Calculation: 
                        a. Create Posts in a day: 10% of DAU => 10% * 500M => 50M create posts per day.
                B. Follow
                    - Calculation:
                        a. Each user follow one new user every week
                            - 500M / 7 => 71M follow api requests per day.
                C. Like and comment
                    - Calculation:
                        a. Each user perform 3 (like/ comment) per day.
                            - 500M * 3 => 1500M
        2. Read Throughput
            - Read posts
            a. Read Posts in a day: Each user view 100 posts per day => 100 * 500 M => 50B posts read in a day.
    4. Storage Estimation
        1. Post:  
            1. Text Post: 100 KB * 50M (write post/day) * 0.2 (as 20%) => 1000 million KB => 1 billion KB => 1TB/day
            2. Image Posts: 0.5 MB * 50 M * 0.6 => 15 million MB => 15 TB/day
            3. Video Posts: 20 MB * 50 M * 0.2 => 200 million MB => 200 TB/day
            4. Total storage Per day: 200 + 15 + 1 => 216 TB/day
            5. Total storage per year: 216 * 365 TB/day => 75 PB
        2. Follow/Unfollow
            1. Assume If 1 follow takes 16 bytes (8 characters) and each user follow 1 new user per week.
            2. 1 day storage: (500M / 7) * 16 bytes => 71.4 million * 16 bytes =>1142 MB => 1 GB
        3. Like and comment
            1. Assuem if each user does 3 comments and each activity takes 216 bytes.
            2. 500 M * 3 * 216 bytes => 324 GB/day
    5. Cache Memory Size
        1. Amount of cache required everyday?
            - 1% of the daily storage, 0.01 * 216 TB = 2 TB
    6. Network/ Bandwidth Estimation 
        1. Our data stored in a day is 216 TB
            1. We can multiply this number to [1.2 - 2.0], this is the range.
            2. Ingress/ per second: 216 * 1.2 / (24 * 60 * 60)
            3. Outgress/ per second:
                - Average size of post: (0.2 * 100KB) + (0.6 * 0.5MB) +(0.2 * 20MB) =4.32MB
                - 50 billion read requests per day * Average size of post =216 PB/day
                - Outgess / per second: 50 billion read requests per day * Average size of post / (24 * 60 * 60) =  2.5 TB/ day
    7. API Design
        


+------------------------+--------------+--------------+--------------+
|                        | VIDEO POSTS  | IMAGE POSTS  | TEXT POSTS   |
+------------------------+--------------+--------------+--------------+
| Avg. Size of Post      | 20 MB        | 0.5 MB       | 100 KB       |
| % of Each Type         | 20%          | 60%          | 20%          |
| Data Storage per Day   | 200 TB       | 15 TB/day    | 1 TB/day     |
+------------------------+--------------+--------------+--------------+

VIDEO POSTS
IMAGE POSTS
TEXT POSTS
Average size
of the post
% of each type
of post
Data Storage
per post
1x
1:11/2:03
"
20 MB
0.5 MB
100 KB
20%
60%
20%
0.2 * 50Million
* 100 KB

```
