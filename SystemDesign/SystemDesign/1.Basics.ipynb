# 1. Basics

```python
# System Design
System Design is the process of defining the architecture, components, modules, interfaces, and data flow of a system to satisfy specified requirements (scalability, latency, availability, etc.).
- It includes
  - HLD: for architecture and components
  - LLD for Interface, class design, patterns
```

```python
# Importance of System Design
1. Scalability and Reliability: Ensures system handles millions of users without failure.
2. Trade-Offs: To Balance Scalability, Cost, Speed, Availability.
```

```python
# Concepts
	•	Latency vs Throughput
	•	CAP Theorem (Consistency, Availability, Partition Tolerance)
	•	Load Balancer, Caching, Database Indexing
	•	Horizontal vs Vertical Scaling
	•	Monolith vs Microservices

# Components
	•	Load Balancers (Nginx, HAProxy)
	•	API Gateway
	•	Web Servers (Apache, Node.js)
	•	App Servers (Docker, Kubernetes)
	•	Relational DB (MySQL, PostgreSQL)
	•	NoSQL DB (MongoDB, Cassandra, Redis)
	•	Caching (Redis, Memcached)
	•	Message Queues (Kafka, RabbitMQ, SQS)
	•	CDN (Cloudflare, Akamai)
	•	Blob Storage (S3, GCS)
	•	Authentication (OAuth, JWT, SSO)
```

```python
# Structured approach to any design problem

Follow the FAIR framework:

F - Functional Requirements
	•	What should the system do? (features)

A - Assumptions & API
	•	Traffic? Users? Data size?
	•	Define APIs: input/output & contracts

I - Interface Design & Storage
	•	Database schema, blob storage, file formats

R - Reliability, Scalability, Tradeoffs
	•	Replication, sharding, availability
	•	Bottlenecks and mitigation
```
