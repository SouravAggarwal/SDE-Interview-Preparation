# 1. uv

```python
- List packages        : uv pip list
- Exact versions       : uv pip freeze
- Package details      : uv pip show <pkg>
- Compare .lock with installed packages: uv pip freeze | diff - uv.lock
- Install by adding the package to .lock file
    - uv add tiktoken
    - uv sync
- Create environment from .lock file
    - Create virtual environment: uv venv --python 3.10 
    - Activate virtual environment: source .venv/bin/activate
    - Install packages from .lock: uv sync
- To run a python file: uv run abc.py
- Find all environements: find . -type d \( -name ".venv" -o -name "venv" -o -name "env" \) -prune 
```
