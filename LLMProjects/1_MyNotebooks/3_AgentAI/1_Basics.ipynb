```python
An LLM agent run tools in a loop to achieve a goal. Although, an LLM model can only generate tokens that describe a tool call.
AI solution that involves any/ all of these:
    1. Multiple LLM calls
    2. LLMs with ability to use Tools
    3. An environment where LLMs interact
    4. A Planner to coordinate activities
    5. Autonomy

Common features:
    • Memory / persistence
    • Planning capabilities
    • Autonomy
    • LLM orchestration via Tools    


The application executes the tool call.

LLM cannot: run code, call api, access files, execute tools.

LLM = planner (language)
App = executor (code)
```

```python
# Agentic Systems
1. Workflows: These are systems where LLMs and tools are orchestrated through predefined code path.
    Examples:
        a. Prompt Chaining
            - You provide the steps that needs to be followed.
        b. Routing
            - WebAPI to handle, the request will be processed by which LLM. (Process is hard coded)
        c. Evaluator-optimizer
            - Technique to ensure high level accuracy in LLM
            - To ensure, LLM provides good answers to customers and does not reveal internal information.
            - The LLM answers are provided to another LLM (Evaluator), to validate if it follows the company guidelines, and then the (Rejection + feedback) is provided to initial generator LLM.
            
2. Agents: Agents are systems where LLMs dynamically direct their own processes and tool usage, maintaining control over how they accomplish task.
    Examples:
        a. Open-ended
            - There is no one particular pattern, they can decide how they will going to colaborate.
        b. Feedback loops
            - Agent can callback itself, untill the agents are satisfied.
        c. No fixd path

```

```python
# Agents Tool Calling
- You generate tool APIs and function and provide access to LLM to call/ excute those functions.
- LLM does not directly call those tools, it return a callback and your primay code executes those tools.
```

# 2. Agent Architecture

```python
- People usually define agents based on what kind of responsibilites different humans will have, which is not the correct way.
- Focus on business objective, 
    1. Start with just one LLM doing everything.
    2. and then divide the task into multiple LLM calls to achieve the business objective.
```

```python
# Components for an Agnetic AI solution
1. Breaking a larger problem into smaller steps carried out by individual processes / models
2. Using Tools / Function Calling / Structured Outputs
3. An Agent Environment in which Agents can collaborate
4. A Planning Agent that coordinates activities
5. Autonomy & Memory - existing beyond a chat with a human
```

```python
# Risks with Agents
1. Unpredictable path
2. Unpredictable output
3. Unpredictable costs.

# What can we do to mitigate Risks
1. Monitor
2. Guardrails ensure your agents behave safely, consistently and within your intended boundaries.
```
