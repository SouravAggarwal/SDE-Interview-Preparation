```python
class Node():
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

N = 3
arrary = []

for i in range(1, N+1):
    root = Node(i)
    arrary.append(PrintTree(root, i))


def PrintTree(root, i):
    
    
    if (root.value < i < 1):
        root.left = PrintTree(root.left, i-1)

   if (N > i > root.value):
        root.right = PrintTree(root.right, i+1)        
```

```python
def find_two_unique_numbers(arr):
    xor_value = 0
    right_most_set_bit_index = 0
    a = b = 0 # buckets
    for num in arr:
        xor_value ^= num

    # find right most set bit
    for i in range(0,32):
        if (xor_value & (1<<i)) > 0:
            right_most_set_bit_index = i
            break
            
    # split numbers into 2 buckets
    for num in arr:
        if num & (1<< right_most_set_bit_index) > 0:
            a ^= num
        else:
            b ^= num
    return (a,b)

find_two_unique_numbers([4, 7, 9, 4, 2, 9, 2, 2])
```
