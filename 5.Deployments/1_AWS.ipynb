# Theory

```python
# Cloud Deployment Archetypes
1. Traditional Cloud Servers - Infrastructure as a Service (IaaS)
    - Where everything started; rent a server and install everything
    - Example: AWS EC2

2. Platform as a Service - (PaaS)
    - Just bring you code; the deployment is handled for you
    - Example: Heroku, Vercel, AWS Beanstalk

3. Container as a Service - (Caas)
    - You provide an app in a container; the service takes care of the rest
    - Example: AWS AppRunner

4. Container Orchestration
    - Run your own fleet of containers and manage how they scale
    - To Manager large number of containers
    - Example: Kubernatics, AWS ECS

5. Serverless Functions - Functions as a Service (FaaS)
    - Upload individual tiny functions and pay per request
    - Pay only for the time function is running
    - It scales down to zero and can handle burst load
    - Example: AWS Lambda
```

# 4. Lambda

```python
Lambda: Individual functions on the cloud, pay for CPU time.
Problem: 
    - Coldstart for the first request.
    - Fixed duration: 15m
    -  DDOS attack: Infinite lambda functions and 
    -  Vendor lock-in: you have to write code according to AWS.
    -  Statefull is not possible.
    - If we connect directly with DB, then no. of DB connections are limited.
            -  To manage this, you need a proxy stateful connection (like  EC2).
        
Pros:
    -  It can scale to infinite by itself, we don't have to manage anything.
        
# How to Run Service Task on Lambda
- Supports Light Weight Service Tasks

EventBridge (cron)/ API request/ SQS message/ S3 file upload
        ↓
     Lambda
        ↓
  Job completes & exits

Service Task on Lambda Use When:
    - Lightweight Service Task
    - Job < 15 minutes
	- Moderate memory/CPU



# Setup Lambda functions
https://github.com/ed-donner/production/blob/main/week2/day2.md

# Pros:
    - Auto-scaling, can send 1000 request (parallely)

# Cons:
    - First request → slower startup
    - Long-running jobs (>15 min)

# Code
def lambda_handler(event, context):
    return {
        "statusCode": 200,
        "body": "Hello from Lambda!"
    }
event   → input data (request, message, etc.)
context → metadata (timeout, request ID, etc.)


# Lambda Internals
  - Lambda uses execution environments (containers), not full servers.
  - Process:
        Request comes in
          ↓
        AWS assigns an execution environment (container)
          ↓
        Your function runs inside it
          ↓
        Environment may be reused for future requests
```

```python
# Use Lambda Layers to install python packages for AWS lambda function.
# This can support multiple lambda functions

Step 1: Create layers directory
    mkdir -p layer/python
    pip install -r requirements.txt -t layer/python

Step 2: Zip the layer
    cd layer
    zip -r layer.zip python

Step 3: In AWS console, Create Lambda Layer
    Lambda → Layers → Create layer
    Upload layer.zip
    Runtime: Python 3.x

Step 4: In your lambda function add that existing layer.
```

# 10. Terraform

```python
Complete Terraform code: https://github.com/SouravAggarwal/LLM-Deployment/blob/main/week2/day4.md

- Create AWS services through code
- Infrastructure as a Code

Features:
1. Controlled: Checked into Git: .tf files with your configuration. So you can review previous configurations at any point of time.
2. Automated: Configures everything, No more AWS console (except IAM)
3. Repeatable: Destroy and Recreate

Terminology:
1. Provide: A vendor like AWS, a plugin to make Terraform apply to that provider.
2. Variable (Parameters): Configure settings that affect your deployment
3. Resource: The building block, Each AWS service to be created is described with a resource.
4. State (Record of resource): Maps the configuration to actual resource. Like what is current state of AWS, what is currently deployed at the moment.
5. Output (Result of deployment): Like actual CloudFront distribution URLs
6. Workspace (Separated State): Its like Isolated state for the same configuration. (Like one box with all configuration for UAT env, and another for production).
    
    
```

# 11. Amazon Sagemaker

Features:
1. End-to-end ML development: Build, Train, fine-tune, and deploy machine learning models.
2. Model training + hosting: Managed infrastructure for running training jobs and deploying trained models as scalable, production-ready endpoints.
3. Designed for MLOps: Includes experiment tracking, model versioning, automated model tuning, and monitoring tools to make production ML repeatable and reliable.

Bedrock vs SageMaker
1. Bedrock:
    Purpose: Use Foundational Models via API
    Models: Pre-trained models from Amazon and partners
    Efforts: Low - just call API, no infra management.
    Best for: Building with frontier models, inference at scale.

2. SageMaker:
    Purpose: Build, Train, fine-tune, and deploy your own ML learning models.
    Models: Any model you build or use open-source models
    Efforts: High - need to manage infra, but you get more control.
    Deploy: Scalable, production-ready endpoints.
    Best for: End to end lifecycle of ML models.
