# 1. Functional Testing

```python
Purpose: Validates that the software functions according to requirements.
When: Done during development, before release, after changes.
```

```python
Type                       | Purpose                                           | When to Use                                    | Benefits
---------------------------|---------------------------------------------------|------------------------------------------------|-------------------------------------------
Unit Testing               | Test smallest code units (functions, classes)     | During development (by developers)            | Catches bugs early, fast feedback
Integration Testing        | Test how different modules interact               | After unit testing, before system testing     | Finds interface issues between modules
System Testing             | Test complete system end-to-end                   | After integration testing                      | Ensures system works as a whole
Smoke Testing              | Quick test to verify major features work          | After build deployment                         | Saves time by catching broken builds
Sanity Testing             | Focused check after minor fixes                   | After bug fix or small change                  | Ensures specific functionality works
Regression Testing         | Verify old features still work after changes      | After any code change                          | Prevents unintended breakage
User Acceptance Testing    | Test by end-users or stakeholders                 | Before production release                      | Confirms system meets business needs
Interface Testing          | Test APIs, UI components for correct communication| After integration                              | Ensures correct data exchange


2. Integration Testing
    - Test cases can be written in code (using frameworks like NUnit, TestNG, JUnit) or defined in test management tools if manual execution is needed.

3. System/ Acceptance Testing
    - Often not written in code directly — instead, they may be described in plain language in tools like Jira, TestRail, or Excel,
    - With automation frameworks (Selenium, Playwright, Cypress, etc.)

4. Smoke Testing
    - Purpose: Quick check to see if the major functionalities are working after a new build/deployment.
    - Done by: QA engineers, CI/CD pipelines.

5. Sanity Testing
    - Purpose: Narrow, focused test to verify a specific functionality after small changes.
    - Done by: QA engineers.

6. Regression Testing
    - Purpose: Verify that recent changes haven’t broken existing features.
    - Mostly has manual regression runs in many companies for UI & exploratory checks.
    - Done by: QA automation engineers + manual QA.

7. User Acceptance Testing (UAT)
    - Purpose: Confirm the product meets business needs before release.
    - Automation is rare — UAT is about validating real-world scenarios.
    - Done by: Business stakeholders, product owners, sometimes QA.        

8. Interface Testing
    - Purpose: Check the interaction between two systems/components (API, UI, DB).
    - Test cases in code?
        - Yes, often automated (API testing scripts, service mocks).
        - Can also be manual (API tools like Postman).
    - Done by: Developers, QA automation engineers.        
```

# 2. Non-Functional Testing

```python
Purpose: Validate how the system works, rather than what it does.
When: Often after functional tests, before going live.
Focus: Non-behavioral aspects (Improves performance, scalability, security)
```

```python
Type                       | Purpose                                           | When to Use                                    | Benefits
---------------------------|---------------------------------------------------|------------------------------------------------|-------------------------------------------
Performance Testing        | Check speed, responsiveness under load            | Before production                              | Ensures good user experience
Load Testing               | Test with expected load                           | Pre-release                                    | Finds performance bottlenecks
Stress Testing             | Push beyond limits                                | Before scaling                                 | Checks system failure behavior
Scalability Testing        | Test ability to handle growth                     | Before growth plans                            | Ensures future readiness
Security Testing           | Find vulnerabilities                              | Before release & periodically                  | Protects data and compliance
Usability Testing          | Check user-friendliness                           | During design & beta                           | Improves UX
Compatibility Testing      | Test on different OS, browsers, devices           | Before release                                 | Ensures broad accessibility
Recovery Testing           | Test ability to recover from crashes              | Before release                                 | Ensures resilience

```

# 3. Other Specialized Testing

```python
Type                       | Purpose                                           | When to Use                                    | Benefits
---------------------------|---------------------------------------------------|------------------------------------------------|-------------------------------------------
Alpha Testing              | Internal user testing                             | Early release                                  | Early feedback
Beta Testing               | Limited real-user release                         | Pre-launch                                     | Real-world validation
A/B Testing                | Compare two versions                              | Feature experiments                            | Data-driven decisions
Exploratory Testing        | Ad-hoc, creative testing                          | Anytime                                        | Finds unexpected bugs
Mutation Testing           | Change code slightly & re-test                    | Advanced QA                                    | Ensures test coverage quality
Accessibility Testing      | Ensure usable by disabled users                   | Pre-release                                    | Compliance, inclusivity
Compliance Testing         | Meet regulations (GDPR, PCI-DSS)                  | Before regulated release                       | Avoids legal issues
Chaos Testing              | Introduce failures to test resilience             | Cloud/Distributed systems                      | Improves fault tolerance


1. Alpha Testing
    - Done in-house by QA and sometimes developers.
    - Controlled environment, often before beta.
    - Example: Internal QA team testing an app before inviting beta testers.

2. Beta Testing
    - Done by external users (selected customers).
    - Real-world usage environment.
    - Example: WhatsApp releasing a beta version to volunteer testers before public release.

3. A/B Testing
    - Often used in product & UI/UX decisions.
    - Example: Showing 50% of users a red "Buy" button, 50% a green one — measuring conversion.

4. Exploratory Testing
    - No formal test scripts.
    - Tester uses creativity, domain knowledge, and intuition.
    - Example: A tester navigates through an app trying edge cases.

5. Mutation Testing
    - For developers.
    - Slightly change source code (e.g., replace > with <) and see if unit tests fail.
    - If they don’t fail → tests are weak.

6. Accessibility Testing
    - Ensure usable by disabled users 
    - Tools: axe, WAVE, NVDA screen reader.
    - Example: Ensuring buttons have proper alt text for screen readers.

7. Compliance Testing
    - Ensures laws/regulations are followed.
    - Example: Payment app following PCI-DSS for credit card storage.

8. Chaos Testing
    - Introduce failures to test resilience
    - Introduced by Netflix (“Chaos Monkey” tool).
    - Randomly shut down servers or disrupt network to check resilience.
    - Goal: system survives unexpected disasters.
```

# 4. Feature Release Strategies

```python
Strategy                 | Purpose
-------------------------|---------------------------------------------------------
Big Bang Release         | Release the entire product/feature set at once to all users.
Phased Rollout           | Release in stages to increasing percentages of users (e.g., 10%, 50%, 100%).
Canary Release           | Release to a small subset of users first to monitor performance before wider rollout.
Dark Launch              | Deploy feature in production but keep it hidden from users until toggled on.
Feature Flag/Toggles     | Control feature availability via configuration switch without redeploying code.
A/B Testing              | Release two versions to different groups to measure performance/impact.
Blue-Green Deployment    | Maintain two identical environments (Blue = live, Green = idle), switch traffic to new version when ready.
Shadow Deployment        | Deploy new version alongside the old one, receive traffic but responses are discarded (used for testing).
Ring Deployment          | Similar to phased rollout but in defined "rings" (internal, beta, pilot, then all).
Pilot Release            | Release to a specific small user group (often paid customers) for validation.


1. Full Deployment
    - Release to all users at once                      
    - Stable, low-risk changes            

2. Big Bang Release
    - What: Everything at once.
    - Pros: Simple, no complex rollout logic.
    - Cons: High risk — if something breaks, all users are affected.
    - Example: Launching a brand-new app version globally.

3. Phased Rollout/ Rolling Deployment
    - What: Gradual increase in users.
    - Pros: Can stop rollout if issues occur.
    - Cons: More operational overhead.
    - Example: Rolling out a new search feature to 10% → 50% → 100%.

4. Canary Release
    - What: Test on a very small group of users first.
    - Pros: Quick issue detection in real environment.
    - Cons: Requires monitoring infrastructure.
    - Example: Deploy to employees only before customers or some users based on geo-location. or device type.

5. Dark Launch
    - What: Deploy feature but keep it hidden.
    - Pros: Lets you test backend without affecting UX.
    - Cons: Can be wasted work if not managed well.
    - Example: Backend supports new recommendation engine but UI hides it.

6. Feature Flags / Toggles
    - What: Enable/disable features dynamically. Hide features behind switches
    - Pros: No redeployment needed to turn on/off. It can be combined with other strategies.
    - Cons: Can create tech debt if flags aren’t cleaned up.
    - Example: Turn on holiday theme UI only in December.

7. A/B Testing
    - What: Show 2 different versions to different set of users to compare.
    - Pros: Data-driven product decisions.
    - Cons: Needs large user base for statistical significance.
    - Example: Different checkout page layouts for conversion testing.  

8. Blue-Green Deployment
    - What: Two identical environments — switch traffic instantly for all users at once, usually via load balancer usually.
    - Pros: Zero downtime, quick rollback if any issue occurs.
    - Cons: Requires double infrastructure.
    - Example: Swap Green (new version) for Blue (old version) instantly.

9. Shadow Deployment
    - What: New version gets real traffic but responses are ignored.
    - Pros: Can test performance and correctness without risk.
    - Cons: Needs routing and duplication logic.
    - Example: API v2 runs alongside API v1 but users still get v1 results.        

10. Ring Deployment
    - Deploy to internal users → small region → all 
    - What: Rollout in predefined groups/rings.
    - Pros: Controlled risk, easy monitoring per ring.
    - Cons: Slightly slower to reach full audience.
    - Example: Internal staff → beta users → small region → all users.

11. Pilot Release
    - What: Early release to a small group of target users.
    - Pros: Business-focused validation.
    - Cons: Feedback can be biased if group isn’t representative.
    - Example: New banking feature only for premium customers first.

```

```python
# Steps in Controlled Rollout
    1.	Internal Testing
    	•	Dev & QA team use staging environment.
    2.	Alpha/Beta Release
    	•	Release to internal or select real users.
    3.	Feature Flagging
    	•	Use a config toggle to turn on for specific users.
    4.	Canary Deployment
    	•	Gradually increase traffic from 1% → 5% → 50% → 100%.
    5.	Monitoring & Observability
    	•	Use dashboards, logs, error tracking.
    6.	Full Rollout
    	•	Enable for everyone if stable.
    7.	Rollback Plan
    	•	Immediate disable or revert if issues found.

# Controlling Which Users Get Which Version
    Step 1: Assign Rollout Bucket
    	•	When user logs in (or on first visit), assign them to a rollout group using:
    	•	hash(userId) % 100 for logged-in users
    	•	Random ID in a cookie for anonymous users
    
    Step 2: Persist
    	•	Store rollout group in:
    	•	User profile in DB (for logged-in)
    	•	Cookie/localStorage (for guest users)
    
    Step 3: Serve Correct Version
    	•	Single build → Serve same HTML/JS, but feature flags determine UI.
    	•	Multiple builds → CDN/server sends correct index.html and assets based on user’s group.

# How to Roll Out to Web App, Mobile App, and Backend
    1. Backend
        •	If the feature changes API contracts or logic significantly, create versioned APIs (e.g., /v2/orders) so old clients don’t break.
    	•	If the change is only additive and backward-compatible, you can reuse the same endpoint with conditional logic (based on feature flags).
    	•	Deploy backend before frontend/mobile changes (so both old and new versions are supported).
    
    2. Web App
    	•	Use feature flags in frontend code to conditionally render UI for the new feature.
    	•	Since web is instantly deployable, you can sync the rollout with backend’s 10% → 50% → 100% stages.
                                             
# Example Workflow for Staged Rollout in Web
    1.	User logs in → Backend checks rollout rule (e.g., hash(user_id) % 100 < rollout_percentage).
    2.	Backend returns feature flags in API response:
            {
              "featureFlags": {
                "newCheckoutFlow": true
              }
            }
    3.	Frontend/mobile checks featureFlags.newCheckoutFlow and shows new UI if true.
    4.	If rollout percentage changes to 50%, backend automatically starts returning true for more users.
```

```python
# Frontend new Feature release in Rolling Deployment
1.  Single Frontend Build + Runtime Feature Flag (Most Common)
	•	One deployed frontend version contains both old and new code paths.
	•	Backend tells the frontend (via API or cookie) whether the user should see the new feature.
	•	Pros: Easy to manage, no multi-build hosting.
	•	Cons: Increases JS bundle size (both versions shipped).
    •	Example:
            if (featureFlags.newDashboard) {
                renderNewDashboard();
            } else {
                renderOldDashboard();
            }

2. Multiple Builds + Routing at CDN/Server (Less Common but Clean Separation)
	•	Create two separate frontend builds:
    	•	/app-v1 → old UI
    	•	/app-v2 → new UI
	•	Use CDN rules or load balancer to send users to the correct build based on:
    	•	Cookie (featureBucket=v2)
    	•	Query param
    	•	User ID hash
	•	Pros: No extra code in bundle, clean separation.
	•	Cons: Two builds to maintain and deploy.
```
