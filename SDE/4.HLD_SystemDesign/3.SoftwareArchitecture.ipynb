```python
# Synchronous vs. Asynchronous Systems
Two systems may communicate with the following ways:
    
Synchronous Communication (Request-Response Model)
    • A request is sent and sender waits for the response before proceeding further. The process is blocked.
    • Blocking calls
    • Tight coupling
    • Example: Traditional HTTP APIs
Asynchronous Communication (Event-Driven Model)
    • A service can send an event and instead of waiting it can continue working on other tasks, and when response arrives it will process the response.
    • Non-blocking
    • Loose Coupling, Decoupled components
    • Example: Message queues & event brokers

```

# 1. Software Architecture

```python
# Software Architecture
    - Software architecture is the high-level structure of a software system. It defines how components interact, how data flows, and how the system behaves under different conditions. 
        It’s like a blueprint for building software.

# Components
1. Components / Modules: The individual parts of the system (e.g., services, classes, functions)
2. Interactions: How components talk (e.g., APIs, events, queues, database access)
3. Data flow: How information moves through the system
4. Deployment view: Where and how the components run (servers, containers, cloud)
5. Non-functional concerns: Performance, security, scalability, reliability, maintainability

# Why it is important
1. Scalability: Ability to handle increased data or traffic
2. Maintainability: How easily the system can be updated or modified
3. Performance: Efficiency and responsiveness under load.

# Architecture Styles
1. Monolithic
2. Multi-Layered (N-tier)
3. Microservice Architecture
4. Event Driven Architecture
5. Client Server Architecture
6. Serverless Architecture
7. Peer-to-Peer (P2P) Architecture
```

```python
# Architecture Styles
1. Monolithic
    - A single-tiered software application where all components are tightly coupled and work as a single unit.
    👍 Pros
        - Easy to develop, manage and deploy
        - No network latency
    👎 Cons
        - Difficult to scale and modify and maintain
        - No fault tolerance
        - A single bug can take down complete application
    - Usecase
        - Leagacy application, small scale apps
        - simple CRUD based app

+----------------------------+
|        Application        |
|  +----------------------+ |
|  |         UI           | |
|  +----------------------+ |
|  |   Business Logic     | |
|  +----------------------+ |
|  | Data Access Layer    | |
|  +----------------------+ |
|          ↓                |
|     +---------+           |
|     | Database |          |
|     +---------+           |
+----------------------------+        


2. Multi-Layered (N-tier)
    - Organizes the application into distinct layers: Presentation, Business Logic, Data Access, and Database. Each layer communicates only with the one 
    - Encapsulates responsibilities, and each layer has a well defined responsibility
    - improves separation of concerns. 
    - Easy to manage and test
    👍 Pros
    	•	Easy to understand
    	•	Separation of concerns
    	•	Reusable and testable components
    	•	Suited for teams with separate UI/backend skills
        •   Easy to scale as compared to Monolithic

    👎 Cons
    	•	Tight coupling between layers
    	•	Can become monolithic
    	•	Slight increase in Latency, due to network hops between tiers
        •   Each layer deploy independently
    - Usecase
        - Banking systems, e-commerce websites
    - Deployment
        - Web Server on Machine A, App Server on Machine B, Database on Machine C.

    User
      ↓
[User Interface] (presentation layer, interacts with user, frontend)
      ↓
[Business Logic] (handles business logic)
      ↓
[Data Access Layer] (access data from DB)
      ↓
[Database]


2.1 Two Tier Architecture
    - Presentation layer has the business logic and it is connected directly with the DB.
    - Ex: webapp connected with DB, Desktop application directly quering the DB.

2.2 Three Tier Architecture
    - UI layer <-> Business Logic Layer <-> DB
    - Better seperation of concern, highly secured, scalable
    

2.3 N-tier architecture
    - Pros
        - It handles High-traffic and complex business logic
        - allows independent scaling.
        - Best scalability, Maintainability, Security, Resilient (Load balancer avoids failed servers, API gateway can retry or cache)
    - Latency consideration
        - More layer = higher latency is not optimized properly.
        - Solution: Caching, Load Balancing
    - Scaling Strategies
        - Vertical scaling: Adding resources to a single server.
        - Horizontal scaling: Adding more servers, load distribution.


                    [ Client / Browser ]
                             │
                             ▼
                    +------------------+
                    |   API Gateway    |  ← Handles auth, routing, rate limiting
                    +------------------+
                             │
                             ▼
                    +------------------+
                    | Load Balancer 🔄 |  ← Distributes requests to app servers
                    +------------------+
                        /       |       \
                       ▼        ▼        ▼
               +---------+ +---------+ +---------+
               | Server1 | | Server2 | | Server3 |   ← App servers
               +---------+ +---------+ +---------+
                    │         │          │
                    ▼         ▼          ▼
              +--------------------------------+
              |         Shared Database        |
              |     (e.g., PostgreSQL/MySQL)   |
              +--------------------------------+
```

### Layered Architecture vs Tiered Architecture

| Aspect              | Layered Architecture                                  | Tiered Architecture                                   |
|---------------------|-------------------------------------------------------|--------------------------------------------------------|
| **Definition**       | Logical separation of concerns in code               | Physical separation across different machines/networks |
| **Focus**            | **How** code is organized                            | **Where** components are deployed                      |
| **Structure**        | Layers such as UI, Business, Data Access in code     | Tiers such as Presentation, Application, Database      |
| **Deployment**       | All layers usually deployed in a single process      | Tiers can be deployed on separate servers              |
| **Communication**    | Function/method calls within a single application    | Network communication (e.g., HTTP, RPC)                |
| **Example**          | MVC structure in a Spring Boot app                   | React frontend → Spring Boot backend → MySQL DB        |
| **Scalability**      | Limited; scale entire app                            | Better; scale each tier independently                  |
| **Use Case**         | Simple or monolithic apps                            | Distributed or enterprise-grade systems                |

```python
3. Microservice Architecture
    - Breaks application into small, independent services, each responsible for a specific business function and communicating via HTTP/gRPC/Queue.
    - Highly scalable and independently deployable units of functionality.
    - It requires: API Gateway, Load Balancing.
    - Each service may have its own database.
    - Pros
        - Scalability
        - Faster development, independent teams can work on different services.
        - Technology agnostic, as each service may have its own tech stack
        - Fault isolation
        - Independent deployment
        - Loosly coupled and modular
        - 
    - Cons
        - Complex orchestration
        - Network latency, inter-service communication
        - Require strong DevOps, deployment and monitoring overhead
    - Usecase
        - Large scale system (amazon, netflix)
        - System needs independent scaling

When to Use Microservices
    - Business Capabilities: Each service should align with a specific business function. (eg: Payment, Order, User)
    - SRP: A microservice should do one thing well.
    - Data ownership: Each service may own its own DB, DB sharing is usually avoided.
    - Independent Deployable: Should be deploy and scale independently without affecting others.
        

How to Structure Microservices?
    - Decompose by Business Domain: Use Domain-Driven Design (DDD) to group services logically.
    - Define Clear APls: Services should communicate via well- defined APls (REST, gRPC, or events).
    - Choose the Right Granularity: Avoid making microservices too large (monolith-in-disguise) or too small (high complexity).
    - Implement Observability: Use logging, monitoring, and tracing to track service interactions.

Communication in Microservices?
Synchronous (which require immediate response)
    - Rest APIs (simple, highly popular, high latency)
    - gRPC (better performance)
Asynchronous (which does not require immediate response)
    - Event Driven Messages (kafka queues, rabitMQ)


Challenges?
    - Data Consistency: Distributed Database and it uses Eventual consistency
    - Distributed Tracing: Difficult to track and debug.
    - Network overhead: More API call, increased latency.
    - Security: Every service needs Authentication, Authorization, and data protection.


Scaling Strategies
    - Horizontal Scaling: Add more instances of a service
    - Auto-scaling: Scale up/down automatically
    - Sharding & Database: Scaling: Split databases for high-traffic services


                                                   
                     +----------------------+
                     |      Client App      |
                     |  (Browser/Mobile)    |
                     +----------+-----------+
                                |
                                v
                    +------------------------+
                    |      API Gateway       |
                    | (Authentication,       |
                    |  Routing, Throttling)  |
                    +-----+---+---+----------+
                          |   |   |
        ------------------    |    ---------------------
       |                      |                      |
       v                      v                      v
+--------------+      +---------------+      +-----------------+
| User Service |      | Order Service |      | Inventory Service |
|  (REST/JSON) |      |   (REST/JSON) |      |    (REST/JSON)    |
+------+-------+      +-------+-------+      +---------+---------+
       |                      |                         |
       v                      v                         v
+--------------+     +----------------+        +-------------------+
| User DB      |     | Order DB       |        | Inventory DB       |
| (PostgreSQL) |     | (MySQL)        |        | (MongoDB)          |
+--------------+     +----------------+        +-------------------+

        ⇅                                ⇅
     Service Discovery              Centralized Logging
      (e.g., Consul, etcd)         (e.g., ELK, Grafana)

          ⇅
     Configuration Service
     (e.g., Spring Cloud Config)
```

```python
4. Event Driven Architecture
    - A system where components communicate through events (messages) instead of direct calls, enabling loose coupling.
    - Components communicate by sending and receiving events via a broker or queue.
    - Decouples producers and consumers; enhances scalability and responsiveness.
    - Pros:
        - Highly decoupled architecture, means services not needs to be aware of each other.
        - Best for Asynchronous workflows.
        - Highly scalable and Loose coupling
        - 
    - Cons
        - Complex debugging, requires good logging.
        - Event ordering challenges
        - Difficult to ensure data consistency across services.
    - Use cases
        - Real-time analytics, notifications/ alerts, stock trading platforms, fraud detection, IoT systems, time taking tasks
        - Logging
        - Real-time Notification 
        - 

Event Producer                                                       Event Consumer            
+-------------+       A        +-----------------+        A        +----------------------+
| Producer 1  |--------------->|                 |--------------->|     Consumer 1       |
|             |                |                 |                | Subscribed to Event A|
+-------------+                |                 |                +----------------------+
                               |                 |
+-------------+       B        |   Event Broker  |       A + B    +----------------------+
| Producer 2  |--------------->|                 |--------------->|     Consumer 2       |
|             |                |                 |                | Subscribed to A + B  |
+-------------+                |                 |                +----------------------+
                               |                 |        
                               |                 |        B        +----------------------+
                               +-----------------+--------------->|     Consumer 3       |
                                                                  | Subscribed to Event B|
                                                                  +----------------------+   

Components:
    1. Event: A message or notification that “something has happened”
    2. Producer: Component that generates/publishes events
    3. Consumer: Component that subscribes/reacts to those events
    4. Broker: Middleware that routes events between producers and consumers (kafka)
    5. Event Bus: Logical channel for event transmission
    6. Event Storage (optional): Log based stored (Elastic)


Types:
    1. Message Queue-Based Asynchronous Communication
        - To communicate b/w multiple service via queue.
                 [Service A: Order Service]
                |
                | sends message (task)
                v
        +----------------------+
        |     Message Queue    |  ← decouples services
        +----------------------+
                |
                v
    [Service B: Notification Service]
         (reads and processes task)


            
    2. Publisher–Subscriber Model (Pub/Sub)
        - In Pub/Sub, publishers emit messages to a channel/topic, and subscribers receive messages from that channel.
        - Features
            •	One-to-many messaging
        	•	Subscribers receive only future messages
        	•	Message is typically removed after delivery
        	•	Subscribers are often decoupled from each other
        - Example:
            - Notification, Chat group, e-commerce order
        - It is Best for Real time notification, sending notifications to multiple users.
        - Rabit MQ, Google and Redis Pub/Sub
            
            [Publisher]
             |
             v
         +--------+
         | Topic  |  ← acts as message broker
         +--------+
            /   \
           /     \
     [Subscriber] [Subscriber]


    2. Event Streaming
        - a continuous stream of events is logged, persisted, and can be replayed by consumers at any time.
        - Key Features
        	•	Persistent (long-lived, replayable) Events are persisted (not deleted after consumption)
        	•	Consumers can replay old events from any point
        	•	Events are stored and consumed in order
        	•	Ideal for high-throughput, real-time data pipelines
        - Example:
            - Kafka
        - Usecase:
            - Logging

Challenges
• Eventual Consistency (No immediate data synchronization), temporary data might be inconsistant (like updating order count) but eventually it will be consistant.
• Ordering Guarantees (Ensuring events are processed in sequence)
• Fault Tolerance & Retries (Handling failures gracefully)
• Debugging Complexity (Tracing events across microservices)        

Best Practices:                     
• Use idempotent event processing to avoid duplicates. Even if a duplicate event is processed, then the output should not be changed.
• Implement dead-letter queues for failed messages: A DLQ is a secondary queue where unprocessed or failed messages are moved after a certain number of retries or error conditions.
• Choose the right event broker based on system needs
• Ensure event versioning to handle schema changes                     
```

```python
5. Client Server Architecture
    - Divides system into two main parts: client (frontend) and server (backend).
    - [Client (browser/app)] ←→ [Server (APIs + DB)]
    👍 Pros:
        •   Centralized control
    	•	Simple and scalable
    	•	Secure and maintainable
    👎 Cons:
    	•	Server bottlenecks
    	•	Not fault tolerant
    	•	Tight client-server dependency
    - Usecase
        Web apps, desktop apps, mobile apps with server backends

6. Serverless Architecture
    - You write functions, and the cloud (like AWS Lambda) handles provisioning, scaling, and infrastructure.
    - No server management; ideal for unpredictable workloads.
    👍 Pros:
    	•	No server management
    	•	Auto-scaling
    	•	Cost-efficient for sporadic traffic

    👎 Cons:
    	•	Cold starts
    	•	Vendor dependency
    	•	Limited runtime environment
    🧩 Use cases:
    	•	Scheduled jobs, image processing, IoT, APIs with low traffic

[Client] → [API Gateway] → [Lambda Function] → [DB/S3]

7. Peer-to-Peer (P2P) Architecture
    - Each node (peer) acts as both a client and a server.
    - Decentralized resource sharing.
    👍 Pros:
    	•	Decentralized
    	•	Resilient to failures
    	•	Scalable
    
    👎 Cons:
    	•	Security concerns
    	•	Data consistency challenges
    	•	No centralized management
    - Usecase
        File sharing (BitTorrent), blockchain, messaging apps
            
[Peer A] ↔ [Peer B] ↔ [Peer C]
        
```
