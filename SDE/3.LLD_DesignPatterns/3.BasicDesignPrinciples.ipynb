# 1. SOLID Principles

```python
- SOLID is an acronym for five design principles intended to make software designs more understandable, flexible, and maintainable.
1. S - Single Responsibility Principle: One class = One job
2. O - Open/Closed Principle: Open for extension, closed for modification
3. L - Liskov Substitution Principle: Subclasses should behave like parent
4. I - Interface Segregation Principle: No client should depend on unused methods
5. D - Dependency Inversion Principle: Depend on abstractions, not concretions

1. Single Responsibility Principle(S)
  - A class should have only one reason to change, means it should have only one responsibility.
  - Ex: A user class should handle business logic only, saving to DB should be done by a seperate class UserRepository.
     - Sending Email, should be done by another class.
      
2. Open/Closed Principle(O)
  - Software entities should be open for extension but closed for modification.
  - Adding new features are closed, changing in existing code/ already tested code is not good.
  - Achieved by Inheritance/ Composition.
# Bad Code                                                             
class Shape:
    def draw(self, shape_type):
        if shape_type == "circle":
            print("Drawing Circle")
        elif shape_type == "rectangle":
            print("Drawing Rectangle")

# Good Code
class Shape:
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("Drawing Circle")

class Rectangle(Shape):
    def draw(self):
        print("Drawing Rectangle")

def render(shape: Shape):
    shape.draw()

3. LSP(L)
    - A child class should be substitutable anywhere its parent class is expected — without breaking behavior.
    - If class S is a subclass of class T, then objects of type T can be replaced with objects of type S without changing the behavior of the program.
    - Objects of a superclass shall be replaceable with objects of its subclasses without breaking the application.
      
# Bad Code
class Bird:
    def fly(self):
        pass

class Ostrich(Bird):
    def fly(self):
        raise Exception("I can't fly")

# Good Code
class Bird:
    pass

class FlyingBird(Bird):
    def fly(self):
        pass

class Sparrow(FlyingBird):
    def fly(self):
        print("Sparrow flying")

class Ostrich(Bird):
    pass  # doesn't fly

# Another Java Example
# Example code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/SOLIDPrinciples/LSP/GoodCode/Main.java
public class WriteableFiles extends ReadableFiles implements IWriteableFiles{
    @override
    pubic void write() { }
}
- Child classes that are readable and writeable will extend Writable class
-  Child classes that are readable only will extend ReadableFiles class 
```

# 2. Design Patterns

```python
- Design patterns are general, proven, reusable solutions to common problems in software design. They provide a template for solving a problem in various situations.
- These are blueprints that can be applied as per use case.
- Focus on Object Interaction, how classes communicate with each other.
- Analogy: Blueprint of a House with standard design.
Benefits:
    - Code Reusability
    - Maintainability: Clean and structured code
    - Understandability: Communication/ common language b/w developers
    - Scalaility
    - Efficiency

There are three types of Design patterns
1. Behavioral DP: Focus on Object interaction, Communication logic b/w objects and Decoupling b/w objects.
    - Ex: Strategy, Observer, Command
2. Creational DP: Focus on Object creation
    - Ex: Singleton, Factory, Builder
3. Structural DP: Object composition
        - Structural design patterns are concerned with how classes and objects are composed to form larger structures.
        - They help ensure that components work together while maintaining flexibility and reusability.
    - Ex: Adapter, Decorator, Composite
```

# 3. Behavioural Design Pattern

```python
Behavioural Design Pattern
 - It focus on how object communicate, interact, managing the flow of information b/w entities.
 - It promotes loose coupling.

1. Momento
    - Capture and restore an object’s internal state without violating encapsulation.
    - Usecase: Undo/Redo functionality, AutoSave/ Rollback.

2. Observer
  - Define a one-to-many dependency so that when one object changes state, all dependents are notified automatically.
  - Usecase: Event listeners, Cricket score display update

3. Strategy
  - Define a family of interchangeable algorithms/behaviors and make them swappable at runtime.
  - Usecase: Discount pricing models in e-commerce, Maps routing algorithm.

4. Command
 - Encapsulate a request as an object, allowing parameterization, queuing, logging, and undoing of operations.
 - Usercase: Where we have a command sender (Remote), and command executor (TV)
          - Remote control

5. Template Method
 - Define the skeleton of an algorithm in a base class but allow subclasses to override some steps.
 - Usecase: File parsing (CSV, JSON, XML with common logic) where readContent logic is different.

6. Iterator
 - Provide a way to access elements of a collection sequentially without exposing its underlying representation.
 - Usecase: Custom collection classes, Traversing lists, trees, graphs, list

7. State
 - Allow an object to change its behavior when its internal state changes (like having multiple “modes”).
 - Usecase: Word doc shows different behaviour in read, write, comment mode.

8. Mediator
 - Define an object that centralizes and coordinates communication between multiple components.
 - Usecase: Chat rooms, Air Traffic Control (all planes communicate to Controler only)

+-------------------+---------------------------+-------------------------------+
| Pattern           | Keyword                   | Use For                       |
+-------------------+---------------------------+-------------------------------+
| Strategy          | Select algorithm          | Runtime flexibility           |
| Observer          | Notify dependents         | Event-driven systems          |
| Command           | Encapsulate request       | Undo/Redo, action history     |
| Chain of Resp.    | Pass request along chain  | Request handling flexibility  |
| Mediator          | Central communication     | Reduce object coupling        |
| Momento           | Save & restore state      | Undo, state snapshots         |
| State             | Change behavior by state  | FSMs, workflow engines        |
| Template Method   | Define skeleton steps     | Reusable algorithm structures |
| Iterator          | Sequential access         | Collections traversal         |
| Visitor           | Add behavior to objects   | Operations on object structure|
| Interpreter       | Parse expressions         | Language or grammar parsing   |
+-------------------+---------------------------+-------------------------------+
```

```python
2.1 Pub/Sub Pattern
    - it also defines how objects interact with each other and how communication flows between them — without being tightly coupled.
    - A messaging pattern where publishers send messages without knowing who the subscribers are.
    - Subscribers listen for events via a broker or message bus.
    - Characteristics:
        - Loosely coupled
        - Asynchronous (can be distributed)
    	- Communication happens via a message broker (Kafka, RabbitMQ, Redis, etc.)
    	- Often used in distributed systems, microservices

Publisher --> [ Message Broker ] --> Subscriber(s)

# Simple Pub/Sub Broker
class MessageBroker:
    def __init__(self):
        self.subscribers = {}

    def subscribe(self, topic, subscriber_fn):
        if topic not in self.subscribers:
            self.subscribers[topic] = []
        self.subscribers[topic].append(subscriber_fn)

    def publish(self, topic, message):
        if topic in self.subscribers:
            for subscriber_fn in self.subscribers[topic]:
                subscriber_fn(message)

# Publisher
class OrderPublisher:
    def __init__(self, broker):
        self.broker = broker

    def place_order(self):
        print("Order placed successfully.")
        self.broker.publish("order_placed", "New order placed!")

# Subscribers
def send_email(message):
    print(f"[Email] {message}")

def send_sms(message):
    print(f"[SMS] {message}")

broker = MessageBroker()
broker.subscribe("order_placed", send_email)
broker.subscribe("order_placed", send_sms)

order_publisher = OrderPublisher(broker)
order_publisher.place_order()

# Output:
# Order placed successfully.
# [Email] New order placed!
# [SMS] New order placed!

Observer vs. Pub/Sub — Comparison Table

Feature                 | Observer Pattern                        | Pub/Sub Pattern
------------------------|------------------------------------------|----------------------------------------------
Coupling                | Tightly coupled                          | Loosely coupled
Communication           | Direct method calls                      | Through a broker or event bus
Scope                   | In-memory, local                         | Distributed, networked
Synchronicity           | Typically synchronous                    | Typically asynchronous
Use Case Example        | GUI event updates                        | Microservices, real-time data feeds
Broker Needed           | No                                       | Yes (e.g., Kafka, Redis, RabbitMQ)
Scalability             | Limited                                  | Highly scalable
Knowledge of Subscribers| Publisher knows subscribers              | Publisher does not know subscribers
Dependency              | Strong dependency between classes        | Decoupled via message channels or topics
```

```python
3. Strategy Pattern

The Strategy Pattern defines a family of algorithms, encapsulates each one, and makes them interchangeable at runtime. Strategy lets the algorithm vary independently from the clients that use it.
Problems:
  - Adding a new payment method require changes in PaymentService.
  - use of if-else conditions to check which payment method is requested.

Benefits:
 - Open/Closed Principle: Add new strategies without changing existing code
 - Flexibility: Strategy can be changed at runtime.
 - No if-else condition to change strategy.

UseCase:	
    •	Payment processors (Card, UPI, Wallet)
	•	Sorting algorithms
	•	Compression (ZIP, RAR, GZIP)
	•	Route planning (shortest path, fastest, least toll)

                +--------------------+
                |     Strategy       |  <--- Interface
                +--------------------+
                | +execute()         |
                +--------------------+
                         ▲
           ----------------------------
           |                          |
+--------------------+    +----------------------+
| ConcreteStrategyA  |    | ConcreteStrategyB    |
+--------------------+    +----------------------+
| +execute()         |    | +execute()           |
+--------------------+    +----------------------+

                ▲
                |
                |
     +-----------------------+
     |     Context           |
     +-----------------------+
     | - strategy: Strategy  |
     +-----------------------+
     | +set_strategy(s)      |
     | +perform_task() to call execute()|
     +-----------------------+

Components:
1. Strategy: Interface that declares execute() or any method for the algorithm.
2. ConcreteStrategyA/B: Actual implementations of different strategies (algorithms).
3. Context: Uses a Strategy object. The behavior is delegated to the strategy instance.

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/BehaviouralPatterns/StrategyPattern/StrategyPattern.java
    
```

# 4. Creational Design Pattern

```python
# Creational Design Pattern
- Creational Design Patterns focus on the creation of objects in a flexible, reusable, and controlled manner, rather than using new keyword or constructors directly.
   - Creational Design Pattern are focused on object creation mechanism, with an aim to optimize the object creation process.

- Goal: They abstract the object initialization process to make system more flexible and re-usable.
- Problem they solve:
  - Prevent tight coupling b/w the code and the object creation logic.
      
1. Singleton DP: Ensure only one instance of a class exists.
  - Logger, DB connection, config loader
  - UseCase: Want to create only 1 instance

2. Builder DP: Construct complex objects step-by-step.
  - HTML/XML builder, creating pizzas/cars step by step
  - UseCase: Object requires multiple steps to construct

3. Prototype DP: Create new objects by cloning existing ones.
 - Cloning UI components, game characters, large object trees
 - UseCase: Need copies of an existing object

4. Factory DP: Delegate object creation to subclasses.
  - UI buttons (Windows, Mac), Notification senders
  - UseCase: Tightly coupled code due to new

5. Abstract Factory DP: Create families of related objects without specifying their concrete classes.
  - Cross-platform UI toolkit, theme factory

+--------------+-------------------------+------------------------------+
| Pattern      | Keyword                 | Use For                      |
+--------------+-------------------------+------------------------------+
| Singleton     | One instance             | Global access, shared state  |
| Factory       | Object creation logic    | Delegated instantiation      |
| Abstract Fact.| Family of factories      | Related objects consistency  |
| Builder       | Step-by-step build       | Complex object construction  |
| Prototype     | Clone object             | Performance, deep copies     |
+--------------+-------------------------+------------------------------+
```

```python
1. Singleton Design Pattern

 - The Singleton Pattern ensures that only one instance of a class is created throughout the lifetime of an application, and provides a global point of access to that instance.
 - Usecase: Database connection, Logger, Configuration Manager

          +---------------------+
          |     Singleton       |
          +---------------------+
          | - static instance   |  << static >>
          | - constructor()     |  << private >>
          +---------------------+
          | +getInstance(): Singleton |
          | +businessMethod()   |
          +---------------------+

Components:
 - Singleton class: Has a private constructor, a static private instance variable, and a static method (getInstance()) to provide global access.

Benefits:
 - One Instance per application.
 - Constructor: Private (not accessible outside the class)
 - Global Access: Via getInstance()
 - Lazy initialization, Initialize only when required.
 - For resource heavy, like DB connection, if multiple instances are created then it will lead to performance issue.
 - Inconsistance State: If there are multiple object of Configuration class, then it may lead to issues.
 - Multiple objects will consume more memory.

Code:
 1. Why we should not have the Singleton class as static?
   - No constructor: You can’t use new, so no instantiation logic 
   - No interfaces
   - No inheritance
   - No lazy initialization
 2. In a multi-threaded env, a synchronized method is used to ensure only one object is created, even if multiple thread try to create object at the same time.
 3. To ensure cloning is not supported, you should override clone() method and throw exception, that cloning is not supported.

Con:  Not thread-safe in multithreaded scenarios without locking

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/CreationalPatterns/SingletonPattern/Solution/AppSettings.java


+---------------------------+-----------------------------+-----------------------------+
|         Feature           |        Singleton Class      |         Static Class        |
+---------------------------+-----------------------------+-----------------------------+
| Instantiation             | Yes (only one instance)     | No instantiation allowed    |
| Object-oriented           | Yes (can inherit, interface)| No (can't inherit/interface)|
| Constructor               | Private constructor         | No constructor (implicit)   |
| State Maintenance         | Yes (can hold state)        | Yes (static state only)     |
| Lazy Initialization       | Yes (can be lazy)           | No (initialized when loaded)|
| Dependency Injection      | Yes                         | No                          |
| Use Case                  | Shared resource with control| Utility/helper methods      |
| Testability               | High (mockable)             | Low (hard to test/mock)     |
| Flexibility               | High                        | Low                         |
| Thread Safety             | Must handle manually        | Handled by runtime          |
+---------------------------+-----------------------------+-----------------------------+
```

```python
2. Factory Design Pattern
 - It helps to centralize the object creation logic and delegate the responsibility of object creation to factory class.
 - The Factory Pattern provides an interface for creating objects in a superclass, but allows subclasses to alter the type of objects that will be created.
 - Usecase:
    - When object creation logic is complex or varies based on input.
    - when want to decouple object creation from usage.
    - Connecting different Databases.
        
Benefits:
1. Encapsulation: Hides object creation logic from the client
2. Open/Closed Principle: Add new product types without modifying existing code
3. Loosely Coupled: Client uses product via interface, not specific class
        

             <<interface>>
             +------------------+
             |     Product      |
             +------------------+
             | +operation()     |
             +------------------+
                    ▲
        -------------------------
        |                       |
+------------------+   +------------------+
|  ConcreteProductA |   | ConcreteProductB |
+------------------+   +------------------+
| +operation()     |   | +operation()     |
+------------------+   +------------------+

             ▲
             |
             |
     +---------------------+
     |     Creator         |  (also called Factory)
     +---------------------+
     | +factoryMethod():Product |  << abstract or interface >>
     +---------------------+

             ▲
        -------------------------
        |                       |
+---------------------+  +----------------------+
| ConcreteCreatorA     |  | ConcreteCreatorB     |
+---------------------+  +----------------------+
| +factoryMethod()     |  | +factoryMethod()     |
+---------------------+  +----------------------+

Components:
1. Product (interface/abstract class): Declares the interface of objects the factory creates.
2. ConcreteProductA/B: Implements the Product interface.
3. Creator (abstract class or interface): Declares factoryMethod() which returns a Product.
4. ConcreteCreatorA/B: Overrides factoryMethod() to instantiate the corresponding ConcreteProduct.

Key Points:
 - The Factory Method pattern creates individual products, while the Abstract Factory pattern is used for creating families of related
          or dependent products through a higher level of abstraction.
 - Factory class actually creates the object.
        
Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/CreationalPatterns/FactoryPattern/Solution/TransportFactory.java
```

```python
3. Abstract Factory Pattern
 - Abstract Factory is a creational design pattern that provides an interface for creating families of related or dependent objects without specifying their concrete classes.
 - Usecase:
    - You need to create multiple related objects (products), but you want to make your system independent of their concrete implementations.

                          +-------------------------+
                          |     AbstractFactory     |<--------------------+
                          +-------------------------+                     |
                          | +createProductA(): A    |                     |
                          | +createProductB(): B    |                     |
                          +-------------------------+                     |
                                   ▲                                        |
                ------------------------------------------                |
                |                                        |                |
+----------------------------+     +----------------------------+        |
|   ConcreteFactory1         |     |   ConcreteFactory2         |        |
+----------------------------+     +----------------------------+        |
| +createProductA(): A1      |     | +createProductA(): A2      |        |
| +createProductB(): B1      |     | +createProductB(): B2      |        |
+----------------------------+     +----------------------------+        |
       |                                     |                            |
       | creates                             | creates                   |
       ▼                                     ▼                            |
+--------------------+           +--------------------+                  |
|   AbstractProductA |<----------|  ConcreteProductA1  |<----------------+
+--------------------+           +--------------------+
                               
+--------------------+           +--------------------+                  
|   AbstractProductB |<----------|  ConcreteProductB1  |                  
+--------------------+           +--------------------+                  


Components
1. AbstractFactory: Interface that declares creation methods for each product type.
2. ConcreteFactory: Implements creation methods to produce concrete products.
3. AbstractProductA/B: Interfaces for products (e.g., Button, Checkbox).
4. ConcreteProductA1/B1: Concrete implementations of those product types.

# Example usecase
1. Product Interface
   - interface Button
   - interface Scrollbar
2. Concrete Products
  - class WindowsButton implements Button
  - class WindowsScrollbar implements Scrollbar
  - class MacOSButton implements Button
  - class MacOSScrollbar implements Scrollbar

3. Abstract Factory Interface
  - interface UIFactory, has abstract methods (button, scrollbar)

4. Concrete Factory Implementation
  - class WindowsFactory implements UIFactory
  - class MacOSFactory implements UIFactory

- With this client code won't be tightly coupled and can change OS, and call methods at runtime.

# Benefits
    - Supports families of related products.
    - Enforces consistency across products (e.g., Mac UI looks like Mac).
    - Makes switching product families easy.

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/CreationalPatterns/AbstractFactoryPattern/Solution/Application.java
```

```python
4. Builder Design Pattern
  - The Builder Pattern is used to construct a complex object step by step. It allows you to build different representations of an object using the same construction process.

Benefits  
  - It seperates the construction of complex object from its representation.
  - The Builder pattern allows setting values during the construction process but once the object is
fully built, it can be made immutable, preventing further modifications.
      
Usecase:
 - When a class object require many parameters, including the optional onces. The constructor become hard to maintain and use.
 - It may has issue like, long constructor parameter list, difficulty to understand.
 - It seperates the initialization of a class/ construction of class object.


Approach-1 
 - Builder can have all the properties of the product.
 - Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/CreationalPatterns/BuilderPattern/Solution/House.java

Apporach-2
 - Builder can have just methods methods and an instance of Product(ie. House), without the need to copy all the attributes.
 - Code: https://medium.com/@vinodkumarbheel61/builder-design-pattern-in-java-a-practical-guide-8a9aaf3d51a3
```

```python
5. Prototype Design Pattern
  - The Prototype Pattern is used when creating a new object by copying an existing object (called the prototype) rather than creating it from scratch. It avoids the cost of repeatedly creating expensive or complex objects.

Usecase
 - When object creation is expensive
 - Cloning capabilities are required. In future, if new properties are added to the class, that should be automatically copied.
 - In a chess game, where you want to save complete board as checkpoint after each move. Manually creating the object of board with all players everytime would be expensive.
    - We want undo/redo functionality with the complete board and players.

Benefits
 - Simply object cloning.
 - Avoids Subclassing: The pattern relies on delegation to the clone method, allowing the sub-class itself to handle object creation, avoiding the need for subclassing.
 - Faster Creation: Avoids heavy initialization
 - Flexibility: No need for subclassing
 - Controlled cloning: Supports deep/shallow copies
 - Each class (ie. chess board and players) provides its own clone method, allowing the entire board to be cloned with its current state.
 - It decouple the cloning an object from its implementation.
 - Consistant in copying.                       

Analogy:
 -  Game development: Create similar enemies/units from a base unit.
 - Word processors: Copy template documents.


         +-------------------+
         |    Prototype<T>   |  <<interface>>
         +-------------------+
         | +clone(): T       |
         +-------------------+
                 ▲
      ------------------------
      |                      |
+-------------------+  +--------------------+
| ConcretePrototype1|  | ConcretePrototype2 |
+-------------------+  +--------------------+
| - fieldA          |  | - fieldX           |  // each class will have its own clone() method.
| +clone()          |  | +clone()           | // clone method returns a new() instance of same class with child objects are also called with copy() (for deep copy)
+-------------------+  +--------------------+

          ▲
          |
          | uses
+--------------------+
|    Client          |
+--------------------+
| - prototype: Proto |
| +operation()       |
+--------------------+

Components
1. Prototype: Interface with clone() method
2. ConcretePrototype: Implements clone to duplicate itself
3. Client: Uses prototype objects to clone new instances
                       
```

# 5. Structural Design Pattern

```python
# Structural Design Pattern

 - Structural design patterns deals with how classes and objects are composed to form larger structures while keeping the system flexible and efficient.
 - They simplify the design of complex system by creating relationships b/w objects.
 - They help in organizing class hierarchies and allow for easy modification and extension.

1. Adapter
2. Decorator
3. Proxy
4. Composite
5. Facade
6. Bridge
7. Flyweightx 

+------------+------------------------+------------------------+
| Pattern    | Keyword                | Use For                |
+------------+------------------------+------------------------+
| Adapter    | Convert interface      | Compatibility          |
| Bridge     | Split interface & impl | Independent variation  |
| Composite  | Tree structure         | Hierarchies            |
| Decorator  | Add functionality      | Dynamic behavior       |
| Facade     | Simplified API         | Complexity hiding      |
| Flyweight  | Object sharing         | Performance            |
| Proxy      | Access control         | Lazy loading, auth     |
+------------+------------------------+------------------------+
```

```python
1. Adapter Pattern
 - Convert the interface of a class into another interface clients expect.
 - Analogy: Electric plug converter to plug a device.
 - Usecase:
    - Connecting legacy code with modern code.
    - Currently we are using a in house email notification, and later we identified using a 3rd party would be cheaper.

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/StructuralPatterns/AdapterPattern/SendGridAdapter.java
```

```python
2. Decorator Pattern
    - Add functionality to objects dynamically.
    - It is used, when we want to add additional values to the base class method/ properties.
        fn Olive(fn Cheese(fn(BasicPizza)))
    - Benefits
        - Flexible, Scalable
        - SRP
        - Open close principle, new toping can be added without changing in existing code.
        - Dynamic change
        - Combinatorial Freedom, toping can be added in any order.

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/StructuralPatterns/DecoratorPattern/Solution/PizzaApp.java
```

```python
3. Proxy Pattern
    - Problem: Sometime direct access to an object is not desirable due to reason such as security, controlled access or resource optimization.
    - Solution: Proxy pattern provides a placeholder for another object to control access to it.
    - Ex: In app if we dont want to load images untill really required.
    - Benefits
        - Lazy loading (loading only when required saving memeory, cpu) and not to load the object again and again (caching)

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/StructuralPatterns/ProxyPattern/Solution/Client.java
```

```python
4. Composite Pattern
    - Idea: Treat individual and composite objects uniformly.
    - Problem: Where we have hierarchies
    - Problem Example: When building systems like a file directory, which consists of both individual files and directories, managing these with standard objects hierarchies might be complex.
    - Soluton: This pattern allows to treat individual and composit object uniformly by representing part-whole heirarchies. 
        - Example: Move() for a file and a directory should be a uniform method. 

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/StructuralPatterns/CompositePattern/Solution/FileSystemApp.java
```

```python
5. Facade Pattern (Facade: The front wall of a large building that you see from outside) 
    - Ex: API Gateway in a microservice 
         - If a client wants to interact with multiple microservices, It would need to communicate directly with all individual services, this would increase the complexity
         - Sol: There is an API gateway, that internally communicates with all the microservices.
    - Analogy: API Gateway acts as a Facade, providing a unified interface to the client while handling all communications with the underlying microservices.
    - Benefits
        - Centralized controll for client and can be used for centralized services (Logging, rate limiting)
        - Intenal classes are hidden from client.                                                         
        - Decoupling, Simplified


Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/StructuralPatterns/FacadePattern/Solution/Client.java
```

```python
6. Flyweight Pattern
    - Use sharing to support large numbers of fine-grained objects efficiently.
    - Mainly for Memory Management (optimization)
    - Example: Icons in google maps (parking, hotel icons)
    - Example Senario: In a shooting game, where multiple identical bullets are fired, Performance can become a issue if each bullet stores the redundant data.
    - Using this pattern, we can reduce memeory overhead by sharing properties of bullet (like appearance), and maintaining unique propeties (like postion and velocity)
        - We divide the properties into 2 parts, (shared properties, and unique properties)

Size comparison for a bullet.
- 4 * 4          * (20 * 20 * 3 * 4) => 5000 bytes
 (x * y postion)   (20 in size with RGB * 4 as integer size) 
- If 5000 bullets are fired, it will take  10^6 bytes space by bullets only.

    - Problem: Creating multiple similar objects in memeory heavy applications can lead to exessive memeory consumption.
    - Solution: By sharing common properties among multiple objects (flyweights), pattern reduces the memeory.
        - All 5000 bullets of type Green colour will share the same object.
        

Code: https://github.com/SouravAggarwal/design-patterns/blob/main/design-patterns/src/main/java/org/prateek/StructuralPatterns/FlyWeightPattern/Solution/Game.java
```
