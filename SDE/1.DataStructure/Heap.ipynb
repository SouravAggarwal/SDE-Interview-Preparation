```python
# Return Kth largest element

def kth_largest(nums, k):
    """ TC: (N * Log K)
      N for for loop and log K for heap size and head operations.
    """
    min_heap = []  # Min-Heap of size K
    for num in nums:
        heapq.heappush(min_heap, num)
        if len(min_heap) > k:
            heapq.heappop(min_heap)  # Remove smallest element
    return heapq.heappop(min_heap)

nums = [3, 2, 1, 5, 6, 4]
print(kth_largest(nums, 2))  # Output: 5 (2nd largest)
```

```python
# Sort a K-sorted array
# A K-sorted array (or nearly sorted array) means that every element is at most K positions away from its true sorted position.

import heapq

def sort_k_sorted_array(arr, k):
    """TC: N log K
    """
    if not arr or k <= 0:
        return arr  # Edge case: No sorting needed

    min_heap = []
    sorted_index = 0

    # Push first K+1 elements into the heap
    for i in range(len(arr)):
        heapq.heappush(min_heap, arr[i])
        if len(min_heap) > k:
            arr[sorted_index] = heapq.heappop(min_heap)
            sorted_index += 1

    # Extract remaining elements from heap
    while min_heap:
        arr[sorted_index] = heapq.heappop(min_heap)
        sorted_index += 1

    return arr

# Example Usage
arr = [6, 5, 3, 2, 8, 10, 9]
k = 3
print(sort_k_sorted_array(arr, k))  # Output: [2, 3, 5, 6, 8, 9, 10]
```

```python
# K closest Number
given number: 7, return K no. of elements nearest to k

arr = [5, 6, 7 ,8, 9]
       -7 -7 -7 -7 -7
    -------------------
      -2 -1 0 1, 2

import heapq

def k_closest_numbers(arr, k, x):
    if not arr or k <= 0:
        return []

    max_heap = []  # Max-Heap to store (-difference, element)

    for num in arr:
        diff = abs(num - x)  # Calculate absolute difference
        heapq.heappush(max_heap, (-diff, -num))  # Store negative to simulate max-heap

        if len(max_heap) > k:
            heapq.heappop(max_heap)  # Remove the farthest element

    return sorted([-num for _, num in max_heap])  # Return sorted results

# Example Usage
arr = [10, 2, 14, 4, 7, 6]
k = 3
x = 5
print(k_closest_numbers(arr, k, x))  # Output: [4, 6, 7]
```

```python
Question 8: Connect ropes to minimize the cost
[3,5, 7, 9, 12]
Cost: When two number are added or two ropes are connect, the sum is added to the total sum.

Idea: Always add 2 smallest numbers at any time.
Steps: 
1. Add all the number to the min heap.
2. At top there will be always smallest numbers on the top.
3. So, top two number and add the result in the same heap.
4. If the result number is smallest, it will remain on the top, else it might go down.
5. In iteration, if size is greater than 2, then pop 2 numbers.


def min_cost_to_connect_ropes(ropes):
    if not ropes:
        return 0

    heapq.heapify(ropes)  # Convert list into Min-Heap in O(N)
    total_cost = 0

    while len(ropes) > 1:
        # Step 1: Take two smallest ropes
        first = heapq.heappop(ropes)
        second = heapq.heappop(ropes)

        # Step 2: Compute cost and add to total
        cost = first + second
        total_cost += cost

        # Step 3: Push new rope back into heap
        heapq.heappush(ropes, cost)

    return total_cost
```

```python
Question 10: Find Median of a Stream of numbers Using Heap
Input: [3, 2, 7, 4]
max Heap.     Min Heap
[2, 4]        [3, 7]

Idea:
    - Max Heap → left half (smaller numbers)
    - Min Heap → right half (larger numbers)
    - Keep both halves balanced and ordered

Solution:
    1) Insert into max heap (as negative in Python)
    2) Move the largest from max heap → min heap
    3) Balance sizes
       - max_heap can have at most 1 extra element

import heapq

class MedianFinder:
    """
    1.	Insertion into heap: O(log N)
	2.	Balancing heaps: O(log N)
	3.	Finding median: O(1)
    """
    def __init__(self):
        self.max_heap = []  # Max-Heap (stores negative values for max behavior)
        self.min_heap = []  # Min-Heap

    def add_num(self, num):
        # Step 1: Push to Max-Heap (left side)
        heapq.heappush(self.max_heap, -num)
        
        # Step 2: Balance by pushing max of left to right heap
        if self.max_heap and self.min_heap and (-self.max_heap[0] > self.min_heap[0]):
            heapq.heappush(self.min_heap, -heapq.heappop(self.max_heap))

        # Step 3: Maintain size property
        if len(self.max_heap) > len(self.min_heap) + 1:
            heapq.heappush(self.min_heap, -heapq.heappop(self.max_heap))
        elif len(self.min_heap) > len(self.max_heap):
            heapq.heappush(self.max_heap, -heapq.heappop(self.min_heap))

    def find_median(self):
        if len(self.max_heap) > len(self.min_heap):
            return -self.max_heap[0]  # Max-Heap root is the median
        return (-self.max_heap[0] + self.min_heap[0]) / 2  # Average of two middle elements

# Example Usage
stream = [5, 15, 1, 3]
median_finder = MedianFinder()

for num in stream:
    median_finder.add_num(num)
    print("Median after inserting", num, ":", median_finder.find_median())

# Expected Output:
# Median after inserting 5 : 5
# Median after inserting 15 : 10.0
# Median after inserting 1 : 5
# Median after inserting 3 : 4.0
```
