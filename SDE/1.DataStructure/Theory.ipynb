```python
# Link List structure cons
-  Linklist and trees are cache inefficient
- Array/ Contigous blocks are cache efficient.

- Arrays are indexed and element can be fetched in O(1) time.
- Tree in array:
  left child index: i * 2 +1
  right child index: i * 2 + 2
  parent: (i-1) / 2

- Array usecase
  -  when we know we need a contigous space and total size is fixed.

- Linklist usecase
  - Dynamic memory allocation
  - Efficient insert/ delete
  - No memory wastage

- In python
   - int and float takes 8 bytes
   - character takes 1 bytes
   - 8 bytes for a pointer 
   - float has size of 16 bytes
```

```python
from collections import deque

# Initialize deque
dq = deque()

# Append elements
dq.append(1)
dq.append(2)

# Append to the left
dq.appendleft(0)

# Pop elements
dq.pop()        # removes 2
dq.popleft()    # removes 0

print(dq)       # Output: deque([1])
```

```python
# Python list Insertion and Deletion.
# deque takes O(1) from both ends, where as List takes O(N)

from collections import deque
a = deque([1, 2, 3, 4, 5])
a.append(6)
a.pop()
a.appendleft(0) # Append at start
a.popleft() # Pop from start

b = [1,2,3,4,5]
b.append(6)
b.pop()
b.insert(0, 10)
b.pop(0)
b
```
