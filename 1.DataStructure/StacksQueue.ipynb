```python
# Stack Problem Identification
Use Stack: To find just smaller value in array by looking up history
- For every element, find the nearest smaller/larger element in the past.
- Find the nearest/next greater/smaller element.

In Stack there are 4 approaches.
 - Stack either in increasing/ decreasing order.
 - and We traverse the array from left to right or right-to-left.

- Store indexes in Stack: when you need Distance / span / Range
- Store values in Stack: when you need greater/ smaller value.

# Algo    
- Ascending order stack: for each element, 
    a. first check if the stack must be in ascending order, pop unnecessary elements with while loop.
    b. Update answer
    c. add current element to the stack.
    
Brute force:
 for i in range(N):
   for j in range(i):
This can be optimized by using a stack.

key point:
 - If stack is not empty, then top element will be the answer.
 - If stack is empty, then -1. 
```

```python

Question: Next greater element for each element

def next_greater_element(arr):
    """Move right to left, keep a stack and remove all smaller elements.
    """
    n = len(arr)
    stack = []
    result = [-1] * n  # Initialize all elements as -1

    # Traverse from right to left
    for i in range(n-1, -1, -1):
        # Remove all elements smaller than arr[i] from the stack
        while stack and stack[-1] <= arr[i]:
            stack.pop()

        # If stack is not empty, the top element is the next greater element
        if stack:
            result[i] = stack[-1]

        # Push the current element into the stack
        stack.append(arr[i])

    return result

# Example
arr = [4, 5, 2, 10, 8]
print(next_greater_element(arr))  # Output: [5, 10, 10, -1, -1




def prev_greater_element(arr):
    """
    Stack: Ascending order stack.
    Move left to right, keep a stack and remove all smaller elements."""
    n = len(arr)
    stack = []
    result = [-1] * n

    for i in range(n):
        while stack and stack[-1] <= arr[i]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result


def next_smaller_element(arr):
    """Move right to left, keep a stack and remove all larger elements."""
    n = len(arr)
    stack = []
    result = [-1] * n

    for i in range(n-1, -1, -1):
        while stack and stack[-1] >= arr[i]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result

def prev_smaller_element(arr):
    """Move left to right, keep a stack and remove all larger elements."""
    n = len(arr)
    stack = []
    result = [-1] * n

    for i in range(n):
        while stack and stack[-1] >= arr[i]:
            stack.pop()

        if stack:
            result[i] = stack[-1]

        stack.append(arr[i])

    return result
```

```python
Question: Remove duplicate characters and return the string in lexographical order.
def remove_duplicates_lexicographically(s):
    """ 
    """
    last_occurrence = {ch: i for i, ch in enumerate(s)}  # Last index of each character
    stack = []  # Stack to store the final characters
    seen = set()  # Set to track elements in stack

    for i, ch in enumerate(s):
        # If character is already in the stack, skip it
        if ch in seen:
            continue

        # Remove elements from the stack if:
        # 1. Stack is not empty
        # 2. The top element is greater than current character (violates lexicographical order)
        # 3. The top element appears later in the string again (can be added back later)
        while stack and stack[-1] > ch and last_occurrence[stack[-1]] > i:
            removed_char = stack.pop()
            seen.remove(removed_char)

        # Add the current character to stack and mark it as seen
        stack.append(ch)
        seen.add(ch)

    return "".join(stack)

# Example
s = "bcabc"
print(remove_duplicates_lexicographically(s))  # Output: "abc"
```

```python
Question: Find largest rectangle area in a histogram,

def largest_rectangle_area(heights):
    """ Calculate
      - Next smaller elements, for default case assume a building at -1 and N index.
    """
    n = len(heights)
    
    # Helper to find Next Smaller to Right
    def next_smaller_right():
        stack = []
        nsr = [n] * n
        for i in range(n - 1, -1, -1):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            if stack:
                nsr[i] = stack[-1]
            stack.append(i)
        return nsr

    # Helper to find Previous Smaller to Left
    def prev_smaller_left():
        stack = []
        psl = [-1] * n
        for i in range(n):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            if stack:
                psl[i] = stack[-1]
            stack.append(i)
        return psl

    # Get boundaries
    nsr = next_smaller_right()
    psl = prev_smaller_left()

    max_area = 0
    for i in range(n):
        width = nsr[i] - psl[i] - 1
        area = heights[i] * width
        max_area = max(max_area, area)

    return max_area

# Example Usage
if __name__ == "__main__":
    histogram = [60, 20, 50, 40, 10, 50, 60]
    print("Maximum Rectangle Area:", largest_rectangle_area(histogram))
```

```python
FIFO: first in first out, ticket counter (person who comes first, will be served first)

Question: Print all binary numbers update N.

from collections import deque

def generate_binary_numbers(n):
    result = []
    q = deque()
    q.append("1")  # Start with 1

    while q:
        binary = q.popleft()
        if len(binary) > n:
            break
        result.append(binary)
        q.append(binary + "0")
        q.append(binary + "1")

    return result
```

```python
# 1. Reverse a link list

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseList(head):
    prev = None
    curr = head
    
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt
        
    return prev


# 2. Detect Cycle in Link list

def hasCycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# 3. Find mid of link list
def middleNode(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow


# 4. Merge 2 sorted list
def mergeTwoLists(l1, l2):
    dummy = curr = ListNode(0)
    while l1 and l2:
        if l1.val < l2.val:
            curr.next = l1
            l1 = l1.next
        else:
            curr.next = l2
            l2 = l2.next
        curr = curr.next
    curr.next = l1 if l1 else l2 # attach remaining nodes.
    return dummy.next


# 5. Remove Nth node from end
def removeNthFromEnd(head, n):
    dummy = ListNode(0, head)
    slow = fast = dummy
    
    for _ in range(n):
        fast = fast.next
    
    while fast.next:
        slow = slow.next
        fast = fast.next
    
    slow.next = slow.next.next # Remove node
    return dummy.next


# 6. Reverse in K Groups 
Input:  1 → 2 → 3 → 4 → 5,  k = 3
Output: 3 → 2 → 1 → 4 → 5

def reverseKGroup(head, k):
    def reverse(head, k):
        prev = None
        curr = head
        while k:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
            k -= 1
        return prev

    count = 0
    curr = head
    while curr and count < k:
        curr = curr.next
        count += 1
    
    if count == k:
        new_head = reverse(head, k)
        head.next = reverseKGroup(curr, k) # After reversing, head becomes the tail So head.next is exactly where you hook the result of the next recursive call.
        return new_head
    
    return head

# 7. Intersection of Two Linked Lists - Intersection is by reference, not value
  Find the node where two linked lists merge. Return null if no intersection.
List A:      4 → 1 ↘
                    8 → 4 → 5
List B:  5 → 6 → 1 ↗

Answer: Node with value 8
Intuition: If you walk both lists and switch to the other list when you hit null, both pointers travel the same total distance (lenA + lenB) and meet at the intersection.

    
def getIntersectionNode(headA, headB):
''' Using 2 points and Pointer switching
'''
    a, b = headA, headB

    while a != b:
        a = a.next if a else headB
        b = b.next if b else headA
    return a  # either intersection node or None    


def getIntersectionNode(headA, headB):
    ''' Using hash set
    '''
    visited = set()

    node = headA
    while node:
        visited.add(node)
        node = node.next

    node = headB
    while node:
        if node in visited:
            return node
        node = node.next

    return None


# 8. Palindrome - Check if LL is palindrome or not

def isPalindrome(head):
    '''Reverse half + compare
    '''
        
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    prev = None
    while slow:
        nxt = slow.next
        slow.next = prev
        prev = slow
        slow = nxt

    left, right = head, prev
    while right:
        if left.val != right.val:
            return False
        left = left.next
        right = right.next
    return True


# 9. Add two number - Link list are already reversed
def addTwoNumbers(l1, l2):
    dummy = curr = ListNode(0)
    carry = 0
    
    while l1 or l2 or carry:
        v1 = l1.val if l1 else 0
        v2 = l2.val if l2 else 0
        
        total = v1 + v2 + carry
        carry = total // 10
        digit = total % 10

        curr.next = ListNode(digit)
        curr = curr.next
        
        l1 = l1.next if l1 else None # No idea
        l2 = l2.next if l2 else None
    
    return dummy.next
```

```python

# 10.    LRU Cache
Problem: Design a cache that, both operation in O(1)
    get(key) → return value if exists, else -1
    put(key, value) → insert/update. If capacity exceeded, evict the Least Recently Used item.

capacity = 2

put(1, 1)   →  [1]
put(2, 2)   →  [1, 2]
get(1)      →  returns 1,  [2, 1]   ← 1 becomes most recent
put(3, 3)   →  evict 2,   [1, 3]   ← 2 was least recent
get(2)      →  returns -1  (evicted)



Apporach: Doubly Link list + HashMap
    HEAD ↔ [LRU] ↔ ... ↔ [MRU] ↔ TAIL
    HEAD and TAIL are dummy sentinel nodes (simplifies edge cases)

- HashMapO(1): access by key
- Doubly Linked: insert/delete anywhere L


from collections import OrderedDict

class Node:
    def __init__(self, key=0, val=0):
        self.key = key
        self.val = val
        self.prev = None
        self.next = None


class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.cache = {}  # key -> Node

        # Dummy sentinel nodes
        self.head = Node()  # LRU side
        self.tail = Node()  # MRU side
        self.head.next = self.tail
        self.tail.prev = self.head

    # ---- Helper: remove a node from the DLL ----
    def _remove(self, node):
        node.prev.next = node.next
        node.next.prev = node.prev

    # ---- Helper: insert a node just before tail (MRU position) ----
    def _insert_at_tail(self, node):
        node.prev = self.tail.prev
        node.next = self.tail
        self.tail.prev.next = node
        self.tail.prev = node

    # ---- GET ----
    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        node = self.cache[key]
        self._remove(node)          # remove from current position
        self._insert_at_tail(node)  # move to MRU
        return node.val

    # ---- PUT ----
    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            self._remove(self.cache[key])  # remove old node

        node = Node(key, value)
        self.cache[key] = node
        self._insert_at_tail(node)         # insert as MRU

        if len(self.cache) > self.capacity:
            # Evict LRU (node just after head)
            lru = self.head.next
            self._remove(lru)
            del self.cache[lru.key]        # ← remove from map too!
```
