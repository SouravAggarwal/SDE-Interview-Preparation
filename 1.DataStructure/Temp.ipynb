```python
def max_path_sum(node):

    def helper(node):
        if node is None:
            return float('-inf'),0

        left_maxsubtree, left_single_path = helper(node.left)
        right_maxsubtree, right_single_path = helper(node.right)        

        path_including_curr_node = node.value +left_single_path + right_single_path
        max_path_sum = max(left_maxsubtree, right_maxsubtree, path_including_curr_node)
        return max_path_sum, node.value + max(left_single_path, right_single_path)
```
