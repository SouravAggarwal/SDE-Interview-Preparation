# 1. Problem Identification

```python
1. Identify Input Type
    If input is:
        - Array/String → Think: sliding window, two pointer, prefix sum
        - Linked list → Think: two pointer, fast-slow pointer
            - LRU: hashMap + Doubly Linked list
        - Tree → Think: DFS / BFS
            - Level Based: BFS
            - Path Based: DFS
            - Need Information from children: Postorder DFS (bottom-up)
            - BST: Inorder traversal gives sorted order
            - Diameter/ Balanced: Return multiple values from recursion
            - Return "Return all unique structures", All combinations: Divide & Conquer + Recursion
                - 
        - Graph → Think: BFS / DFS / Union-Find
            - Shortest Path
                - Unweighted → BFS
                - Weighted → Dijkstra
                - Negative weights → Bellman-Ford
                - DAG → Topological + DP
            - Cycle Detection
                - Undirected → DFS with parent
                - Directed → DFS with recursion stack
            - Connected Components
                - DFS/ Union Find
            - Minimum Spanning Tree
                - Kruskal (Union Find/ Prim (Heap))
            - 

        - Intervals → Think: sorting + merge
        - Matrix → Think: BFS / DFS / backtracking
            - Island problems → DFS/BFS
            - Shortest path grid → BFS
            - All paths → Backtracking
            - Multi-source shortest → BFS with multiple start nodes


2. Identify by Keywords
    - “Subarray”: Sliding window / Prefix sum
    - “Substring”: Sliding window
    - “Sorted array”: Binary search / Two pointer
    - “Kth largest/smallest”: Heap
    - “Top K frequent”: Heap + HashMap
    - “Shortest path”: BFS / Dijkstra
    - “Cycle detection”: Fast-slow / DFS
    - “All combinations”: Backtracking
    - “Max/Min in window”: Monotonic queue
    - “Overlapping intervals”: Sort + Merge
    - “Connected components”: DFS / Union Find
    - “Range Query“: Prefix Sum (for static array), Segment Tree (for Updates Involved)
    - Monotonic Behaviour: Monotonic Stack/ Deque
        - Question: Next greater element, stock span, Largest rectangle, Sliding window

3. Identify by Problem Type
EXISTENCE:
    → Usually DFS / BFS / HashSet

COUNTING:
    → DP / Prefix sum / HashMap

OPTIMIZATION:
    → Greedy / DP / Sliding window

MINIMUM STEPS:
    → BFS

ALL POSSIBILITIES:
    → Backtracking

KTH ELEMENT:
    → Heap / Quickselect

INTERVAL MERGING:
    → Sort + Merge

OVERLAPPING:
    → Sorting
```

```python
3. Decision Flow


1. What is the input?
   - Array/String → Go A
   - Linked List → Go B
   - Tree → Go C
   - Graph → Go D
   - Matrix → Go E

--------------------------------

A: ARRAY / STRING

  Is it sorted?
      YES → Binary Search / Two Pointer
      NO  → Continue

  Is it subarray/substring?
      YES → Sliding Window / Prefix Sum

  Is it K largest/smallest?
      YES → Heap

  Is it combination/permutation?
      YES → Backtracking

  Is it max/min range?
      YES → Monotonic Stack

--------------------------------

B: LINKED LIST

  Cycle? → Fast & Slow Pointer
  Reverse? → Iterative pointer reversal
  Merge? → Two pointer

--------------------------------

C: TREE

  Level order? → BFS
  Depth based? → DFS
  Lowest Common Ancestor? → DFS recursion
  Balanced? → Bottom-up DFS

--------------------------------

D: GRAPH

  Shortest path (unweighted)? → BFS
  Shortest path (weighted)? → Dijkstra
  Detect cycle? → DFS / Union Find
  Components? → DFS / Union Find

--------------------------------

E: MATRIX

  Island type? → DFS/BFS
  Shortest path grid? → BFS
  All paths? → Backtracking
```

```python
Problem Approach Steps:
    1. Understand constraints
    2. Identify structure type
    3. Identify problem family
    4. Try brute force
    5. Improve using pattern triggers
    6. Choose data structure
    7. Write clean code
    8. Analyze complexity
    9. Dry run

Optimization
    Can I:
    - Use HashMap?
    - Use Sorting?
    - Use Binary Search?
    - Use Two Pointer?
    - Use Sliding Window?
    - Use Prefix Sum?
    - Precompute something?
```
