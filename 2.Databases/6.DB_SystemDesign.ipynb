# 1. Twitter System Design

```python
# Post tweets Designs
Approach 1. A user Posts a tweet -> We add the tweet in a queue -> response success returned to user, and later the tweet is saved in a DB.
    - Issue: After posting a tweet, users will see their tweets after sometime, and can not edit in the mean time.
Approach 2: A user Posts a tweet => a direct API call to the server -> and save in DB.
    - Issue: If connection fails from user end, the api call will get cancelled.
    - Issue: Bad UX, as the api call will take time to post the tweet.
Approach 3: Save the user tweet in the app local DB, and make an API to directly save in DB.
    - Pros: Now, if api fails and whenever user comes back online, he will see an error message, post tweet failed, and a retry button.

# Follow People Design
Requirements
    - Follow a new user
    - Get list of all followers/ followees
    - is_follow?, When we visit a new profile, this is displayed in the follow button.
    
        
```
